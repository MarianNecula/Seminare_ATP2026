# Partea I – Stiva

## 1. Definitie

O **stiva** este o structura de date in care lucram dupa regula:

**ultimul intrat, primul iesit**

Aceasta regula se numeste **LIFO**:

**Last In, First Out**

Exemplul cel mai intuitiv este un teanc de farfurii:
- pui o farfurie;
- pui alta deasupra;
- pui inca una deasupra.

Cand iei o farfurie, o iei pe cea de sus, adica pe ultima pusa.

Aceasta este exact ideea de stiva:
- adaugam „sus”;
- scoatem tot „sus”.


## 2. Reprezentare simpla

Daca introducem valorile:
- 10
- 20
- 30

atunci putem reprezenta stiva astfel:

```text
    varf
     |
    [30]
    [20]
    [10]
```

Observatie importanta:
toate operatiile esentiale se fac la **varf**.

Nu scoatem din mijloc.
Nu scoatem de la baza.
Lucram doar in partea de sus.


## 3. Exemplu

In Python, o lista poate fi folosita foarte usor pentru a simula o stiva.

### Exemplu
Adaugam trei valori si scoatem una.


````python
stiva = []

stiva.append(10)
stiva.append(20)
stiva.append(30)

print("Stiva dupa adaugare:", stiva)

x = stiva.pop()
print("Element scos:", x)

print("Stiva ramasa:", stiva)
````

ultimul pus este primul scos.

Acesta este primul model important pe care trebuie sa il retii:
**daca problema cere sa lucram cu cel mai recent element adaugat, stiva este o alegere buna.**


## 5. Operatii fundamentale

O stiva are cateva operatii de baza:

- `push(x)` – adauga un element;
- `pop()` – scoate ultimul element;
- `peek()` – arata ultimul element fara sa il scoata;
- `is_empty()` – verifica daca stiva este goala;
- `size()` – intoarce numarul de elemente.

Aceste operatii apar in aproape toate implementarle de stive, chiar daca numele exacte pot varia putin de la o sursa la alta.


````python
class Stack:
    def __init__(self):
        self.data = []

    def push(self, x):
        self.data.append(x)

    def pop(self):
        return self.data.pop()

    def peek(self):
        return self.data[-1]

    def is_empty(self):
        return len(self.data) == 0

    def size(self):
        return len(self.data)


s = Stack()
s.push(5)
s.push(7)
s.push(9)

print("Varf:", s.peek())
print("Scoatem:", s.pop())
print("Varf dupa scoatere:", s.peek())
print("Dimensiune:", s.size())
````

## 6. Clasa `Stack`

Aceasta clasa grupeaza toate operatiile utile intr-un singur loc.

- `self.data` este locul unde pastram valorile;
- `push` adauga valori noi;
- `pop` scoate valoarea din varf;
- `peek` ne permite sa privim varful fara sa modificam stiva;
- `is_empty` ne spune daca mai exista elemente;
- `size` ne spune cate elemente sunt.

Este important sa intelegi ca o structura de date nu este doar o colectie de valori, ci si un mod bine organizat de a lucra cu ele.

## 11. Problema clasica: inversarea unui cuvant

Stiva este excelenta pentru inversare, deoarece ordinea de iesire este inversa fata de ordinea de intrare.

Daca punem literele unui cuvant in stiva si apoi le extragem, obtinem automat forma inversata.


````python
def inverseaza(text):
    s = Stack()

    for litera in text:
        s.push(litera)

    rezultat = ""

    while not s.is_empty():
        rezultat += s.pop()

    return rezultat


print(inverseaza("python"))
print(inverseaza("student"))
print(inverseaza("algoritm"))
````

## 12. Explicatie pentru inversare

Pentru cuvantul `abc`:
- punem `a`
- punem `b`
- punem `c`

La extragere:
- iese `c`
- apoi `b`
- apoi `a`

Rezultatul este `cba`.

Aceasta este o idee foarte importanta in programare:
uneori nu rezolvam problema „direct”, ci alegem o structura care produce natural comportamentul dorit.

## 14. Problema clasica: verificarea parantezelor

Aceasta este una dintre cele mai cunoscute aplicatii ale stivei.

Exemple:
- `(a+b)` este corecta;
- `((x+y))` este corecta;
- `(()` este gresita;
- `())` este gresita.

De ce functioneaza stiva aici?
Pentru ca fiecare paranteza inchisa trebuie sa se potriveasca cu **cea mai recenta** paranteza deschisa ramasa fara pereche. Aceasta relatie „cea mai recenta” este exact o situatie de tip LIFO.


````python
def paranteze_corecte(text):
    s = Stack()

    for c in text:
        if c == "(":
            s.push(c)

        if c == ")":
            if s.is_empty():
                return False
            s.pop()

    return s.is_empty()


teste = ["(a+b)", "((x+y))", "(()", "())", ""]
for t in teste:
    print(t, "->", paranteze_corecte(t))
````

## 15. Urmarire pas cu pas

Pentru textul `(())`:
1. citim `(` → punem in stiva;
2. citim `(` → punem in stiva;
3. citim `)` → scoatem o paranteza;
4. citim `)` → scoatem inca una.

La final stiva este vida. Prin urmare, expresia este corecta.

Daca ramane ceva in stiva, inseamna ca exista deschideri fara inchidere.  
Daca vrem sa scoatem dintr-o stiva vida, inseamna ca exista o inchidere fara deschidere.

## 17. Aplicatie: Undo

Operatia Undo este o situatie naturala pentru stiva.

De fiecare data cand modificam textul, memoram starea anterioara.
Cand vrem sa anulam ultima operatie, revenim la ultima stare salvata.

Observa ideea:
- ultima modificare
- se anuleaza prima

Exact aceasta ordine face stiva potrivita.


````python
class EditorSimplu:
    def __init__(self):
        self.text = ""
        self.undo = Stack()

    def scrie(self, bucata):
        self.undo.push(self.text)
        self.text += bucata

    def sterge_ultimele(self, k):
        self.undo.push(self.text)
        self.text = self.text[:-k]

    def undo_last(self):
        if not self.undo.is_empty():
            self.text = self.undo.pop()

    def afiseaza(self):
        return self.text


ed = EditorSimplu()
ed.scrie("Ana")
ed.scrie(" are")
ed.scrie(" mere")
print(ed.afiseaza())

ed.sterge_ultimele(5)
print(ed.afiseaza())

ed.undo_last()
print(ed.afiseaza())
````

## 19. Problema: palindrom

Un palindrom este un cuvant care se citeste la fel din ambele directii.

Exemple:
- `radar`
- `level`
- `ana`

Stiva ne ajuta deoarece poate produce usor forma inversa a unui text.


````python
def este_palindrom(text):
    s = Stack()

    for c in text:
        s.push(c)

    invers = ""
    while not s.is_empty():
        invers += s.pop()

    return text == invers


for x in ["radar", "level", "python", "ana"]:
    print(x, "->", este_palindrom(x))
````

# Partea a II-a – Coada

## 1. Definitie

O **coada** functioneaza foarte diferit fata de stiva.

Aici regula este:

**primul intrat, primul iesit**

Aceasta regula se numeste **FIFO**:

**First In, First Out**

Exemplul cel mai natural este o coada de persoane la ghiseu:
- cine vine primul
- este servit primul.


## 2. Reprezentare

Daca adaugam valorile 10, 20, 30, coada poate fi desenata astfel:

```text
fata -> [10] [20] [30] <- final
```

Observa:
- noile valori intra la **final**;
- valorile ies din **fata**.

Aceasta este diferenta esentiala fata de stiva.


## 3. Implementare - operatii

Pentru coada vom avea operatii de tip:
- `enqueue(x)` – adauga la final;
- `dequeue()` – scoate din fata;
- `front()` – arata primul element;
- `empty()` – verifica daca structura este goala.

Este foarte important sa legi aceste idei de analogia reala a unei cozi de persoane.


````python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.prev = None


class Queue:
    def __init__(self):
        self.first = None
        self.last = None

    def enqueue(self, x):
        n = Node(x)

        if self.first is None:
            self.first = n
            self.last = n
        else:
            n.prev = self.last
            self.last.next = n
            self.last = n

    def dequeue(self):
        val = self.first.val
        self.first = self.first.next

        if self.first is not None:
            self.first.prev = None
        else:
            self.last = None

        return val

    def empty(self):
        return self.first is None

    def front(self):
        return self.first.val
````

## 4. Explicatie detaliata pentru coada

### `enqueue(x)`
Adauga un nou element la final.

### `dequeue()`
Scoate elementul din fata.

### `front()`
Arata cine este primul la rand, fara sa il elimine.

### `empty()`
Verifica daca mai exista elemente.

Trebuie sa observi diferenta de comportament:
la stiva lucram cu ultimul element; la coada lucram cu primul element introdus.


````python
q = Queue()

q.enqueue("Ana")
q.enqueue("Bogdan")
q.enqueue("Carmen")

print("Primul la rand:", q.front())
print("Se serveste:", q.dequeue())
print("Se serveste:", q.dequeue())
print("Se serveste:", q.dequeue())
````

## 9. BFS

BFS inseamna **Breadth-First Search**.

La nivel introductiv, putem intelege aceasta idee astfel:
vizitam mai intai nodurile cele mai apropiate de punctul de start, apoi mergem mai departe pe niveluri.

De ce folosim coada?
Pentru ca vrem sa procesam nodurile in ordinea in care au fost descoperite. Exact acest comportament este FIFO.


````python
def bfs(g, start):
    q = Queue()
    q.enqueue(start)

    viz = set([start])

    while not q.empty():
        n = q.dequeue()
        print(n)

        for v in g[n]:
            if v not in viz:
                viz.add(v)
                q.enqueue(v)


graf = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": [],
    "D": []
}

bfs(graf, "A")
````

## 10. Descriere BFS

Pornim cu nodul `A`.

- initial, coada contine doar `A`;
- scoatem `A`, apoi adaugam `B` si `C`;
- scoatem `B`, apoi adaugam `D`;
- scoatem `C`;
- scoatem `D`.

Ordinea este:
```text
A, B, C, D
```

Aceasta ordine exista pentru ca folosim o coada. Daca am folosi o stiva, parcurgerea s-ar comporta diferit.


## 11. Deque – definitie

Uneori vrem o structura mai flexibila, in care sa putem lucra la ambele capete. Aceasta structura se numeste **deque** (se pronunta precum **deck**).

Un deque permite:
- adaugare la inceput;
- adaugare la final;
- scoatere de la inceput;
- scoatere de la final.

Pentru inceput, este suficient sa retii ca deque-ul este „o coada mai flexibila”.


````python
class Deque:
    def __init__(self):
        self.first = None
        self.last = None

    def empty(self):
        return self.first is None

    def push_right(self, x):
        n = Node(x)
        if self.empty():
            self.first = self.last = n
        else:
            n.prev = self.last
            self.last.next = n
            self.last = n

    def push_left(self, x):
        n = Node(x)
        if self.empty():
            self.first = self.last = n
        else:
            n.next = self.first
            self.first.prev = n
            self.first = n

    def pop_left(self):
        val = self.first.val
        self.first = self.first.next
        if self.first is not None:
            self.first.prev = None
        else:
            self.last = None
        return val

    def pop_right(self):
        val = self.last.val
        self.last = self.last.prev
        if self.last is not None:
            self.last.next = None
        else:
            self.first = None
        return val
````

````python
d = Deque()

d.push_right(10)
d.push_right(20)
d.push_left(5)

print(d.pop_left())
print(d.pop_right())
````

## 12. Coada cu prioritate – definitie

Intr-o coada obisnuita, ordinea este data de sosire.

Intr-o coada cu prioritate, unele elemente pot fi tratate mai devreme pentru ca sunt mai importante.

Exemple:
- incident critic;
- bug important;
- cerere obisnuita.

Aceasta structura este foarte utila in software real, unde nu toate task-urile au aceeasi urgenta.


````python
class PriorityQueue:
    def __init__(self):
        self.first = None
        self.last = None

    def empty(self):
        return self.first is None

    def enqueue(self, value, priority):
        n = Node((value, priority))

        if self.empty():
            self.first = self.last = n
            return

        current = self.first

        while current is not None and current.val[1] <= priority:
            current = current.next

        if current is self.first:
            n.next = self.first
            self.first.prev = n
            self.first = n
        elif current is None:
            n.prev = self.last
            self.last.next = n
            self.last = n
        else:
            prev_node = current.prev
            n.prev = prev_node
            n.next = current
            prev_node.next = n
            current.prev = n

    def dequeue(self):
        val = self.first.val
        self.first = self.first.next

        if self.first is not None:
            self.first.prev = None
        else:
            self.last = None

        return val
````

````python
pq = PriorityQueue()

pq.enqueue("cerere normala", 3)
pq.enqueue("incident critic", 1)
pq.enqueue("bug important", 2)

print(pq.dequeue())
print(pq.dequeue())
print(pq.dequeue())
````
