# Algoritmi de sortare


## 1. Notiuni generale despre sortare

Sortarea este operatia prin care elementele unei colectii sunt reordonate dupa o regula, de obicei:
- crescator;
- descrescator.

Exemplu:
```text
[5, 2, 9, 1, 7] -> [1, 2, 5, 7, 9]
```

Sortarea este importanta deoarece simplifica alte operatii:
- cautarea;
- compararea;
- detectarea duplicatelor;
- calculul medianei;
- pregatirea datelor pentru analiza.

### Criterii de comparatie intre algoritmi
Cand evaluam un algoritm de sortare, ne intereseaza:
- timpul de executie;
- memoria suplimentara;
- stabilitatea;
- simplitatea implementarii;
- tipul de date pentru care este potrivit.


## 2. Tabel comparativ

| Algoritm | Timp mediu | Caz nefavorabil | Memorie suplimentara | Stabil | Observatie |
|---|---:|---:|---:|---|---|
| Bubble Sort | O(n²) | O(n²) | O(1) | Da | simplu, dar lent |
| Selection Sort | O(n²) | O(n²) | O(1) | Nu, clasic | putine swapuri |
| Insertion Sort | O(n²) | O(n²) | O(1) | Da | foarte bun pe liste mici |
| Merge Sort | O(n log n) | O(n log n) | O(n) | Da | robust si predictibil |
| Quick Sort | O(n log n) | O(n²) | depinde de implementare | Nu, clasic | foarte rapid in practica |
| Radix Sort | O(nk) | O(nk) | O(n + baza) | Da, daca auxiliarul e stabil | foarte bun pentru numere intregi |


# 3. Bubble Sort

## Idee
Bubble Sort compara elemente adiacente si le inverseaza daca sunt in ordine gresita.  
Dupa fiecare trecere completa, cel mai mare element ajunge la final.

## Exemplu pas cu pas
Lista initiala:
```text
[5, 3, 8, 4, 2]
```

Prima trecere:
- 5 si 3 -> swap -> [3, 5, 8, 4, 2]
- 5 si 8 -> ok
- 8 si 4 -> swap -> [3, 5, 4, 8, 2]
- 8 si 2 -> swap -> [3, 5, 4, 2, 8]

A doua trecere:
- 3 si 5 -> ok
- 5 si 4 -> swap -> [3, 4, 5, 2, 8]
- 5 si 2 -> swap -> [3, 4, 2, 5, 8]

Continuam pana cand lista devine:
```text
[2, 3, 4, 5, 8]
```


````python
def bubble_sort(v):
    n = len(v)
    for i in range(n):
        for j in range(0, n - i - 1):
            if v[j] > v[j + 1]:
                v[j], v[j + 1] = v[j + 1], v[j]
    return v

print(bubble_sort([5, 3, 8, 4, 2]))
````

## Varianta cu afisare dupa fiecare trecere


````python
def bubble_sort_with_steps(v):
    n = len(v)
    for i in range(n):
        for j in range(0, n - i - 1):
            if v[j] > v[j + 1]:
                v[j], v[j + 1] = v[j + 1], v[j]
        print(f"Dupa trecerea {i + 1}: {v}")
    return v

bubble_sort_with_steps([5, 3, 8, 4, 2])
````

## Complexitate
- caz favorabil: O(n), daca folosim oprire anticipata;
- caz mediu: O(n²);
- caz nefavorabil: O(n²);
- memorie: O(1).

## Avantaje
- foarte simplu;
- usor de urmarit;
- stabil.

## Dezavantaje
- lent pe liste mari;
- multe comparatii si swapuri.

## Exercitiu rezolvat
Sortare descrescatoare:


````python
def bubble_sort_desc(v):
    n = len(v)
    for i in range(n):
        for j in range(0, n - i - 1):
            if v[j] < v[j + 1]:
                v[j], v[j + 1] = v[j + 1], v[j]
    return v

print(bubble_sort_desc([5, 3, 8, 4, 2]))
````

# 4. Selection Sort

## Idee
Selection Sort cauta minimul din partea nesortata si il aduce pe prima pozitie disponibila.

## Exemplu pas cu pas
Lista:
```text
[64, 25, 12, 22, 11]
```

Pasul 1:
- minimul este 11
- il punem pe prima pozitie
```text
[11, 25, 12, 22, 64]
```

Pasul 2:
- minimul din rest este 12
```text
[11, 12, 25, 22, 64]
```

Pasul 3:
- minimul din rest este 22
```text
[11, 12, 22, 25, 64]
```


````python
def selection_sort(v):
    n = len(v)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if v[j] < v[min_index]:
                min_index = j
        v[i], v[min_index] = v[min_index], v[i]
    return v

print(selection_sort([64, 25, 12, 22, 11]))
````

## Varianta cu afisarea minimului


````python
def selection_sort_with_min(v):
    n = len(v)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if v[j] < v[min_index]:
                min_index = j
        print(f"Pasul {i + 1}: minimul este {v[min_index]}")
        v[i], v[min_index] = v[min_index], v[i]
        print(f"Lista dupa pasul {i + 1}: {v}")
    return v

selection_sort_with_min([64, 25, 12, 22, 11])
````

## Complexitate
- timp: O(n²) in toate cazurile importante;
- memorie: O(1).

## Avantaje
- simplu;
- numar relativ mic de swapuri;
- potrivit pentru invatare.

## Dezavantaje
- lent pe liste mari;
- nu profita de faptul ca lista este aproape sortata;
- nu este stabil in forma clasica.

## Exercitiu rezolvat
Sortare descrescatoare:


````python
def selection_sort_desc(v):
    n = len(v)
    for i in range(n):
        max_index = i
        for j in range(i + 1, n):
            if v[j] > v[max_index]:
                max_index = j
        v[i], v[max_index] = v[max_index], v[i]
    return v

print(selection_sort_desc([64, 25, 12, 22, 11]))
````

# 5. Insertion Sort

## Idee
Insertion Sort construieste treptat o portiune sortata a listei.

La fiecare pas:
- luam un element nou;
- il inseram in pozitia corecta in partea deja sortata.

Analogia clasica este ordonarea cartilor de joc in mana.

## Exemplu pas cu pas
Lista:
```text
[5, 3, 8, 4, 2]
```

Pasul 1:
- partea sortata: [5]
- luam 3 si il inseram -> [3, 5, 8, 4, 2]

Pasul 2:
- partea sortata: [3, 5]
- 8 ramane la final -> [3, 5, 8, 4, 2]

Pasul 3:
- luam 4 si il inseram intre 3 si 5 -> [3, 4, 5, 8, 2]

Pasul 4:
- luam 2 si il inseram la inceput -> [2, 3, 4, 5, 8]


````python
def insertion_sort(v):
    for i in range(1, len(v)):
        key = v[i]
        j = i - 1

        while j >= 0 and v[j] > key:
            v[j + 1] = v[j]
            j -= 1

        v[j + 1] = key
    return v

print(insertion_sort([5, 3, 8, 4, 2]))
````

## Complexitate
- caz favorabil: O(n), daca lista este deja aproape sortata;
- caz mediu: O(n²);
- caz nefavorabil: O(n²);
- memorie: O(1).

## Avantaje
- simplu;
- stabil;
- foarte bun pe liste mici;
- eficient pe liste aproape sortate.

## Dezavantaje
- lent pe liste mari;
- multe deplasari in caz nefavorabil.

## Exercitiu rezolvat
Numararea deplasarilor:


````python
def insertion_sort_count(v):
    shifts = 0
    for i in range(1, len(v)):
        key = v[i]
        j = i - 1

        while j >= 0 and v[j] > key:
            v[j + 1] = v[j]
            j -= 1
            shifts += 1

        v[j + 1] = key

    return v, shifts

print(insertion_sort_count([9, 4, 7, 3, 1]))
````

# 6. Merge Sort

## Idee
Merge Sort foloseste strategia divide et impera:
1. imparte lista in doua jumatati;
2. sorteaza fiecare jumatate;
3. combina cele doua jumatati sortate.

## Exemplu pas cu pas
Lista:
```text
[38, 27, 43, 3, 9, 82, 10]
```

Impartire:
```text
[38, 27, 43] si [3, 9, 82, 10]
```

Continuam pana la subliste de cate un element, apoi combinam:
- [27] si [43] -> [27, 43]
- [38] si [27, 43] -> [27, 38, 43]
- [3] si [9] -> [3, 9]
- [82] si [10] -> [10, 82]
- apoi combinam totul.


````python
def merge(left, right):
    result = []
    i = 0
    j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

def merge_sort(v):
    if len(v) <= 1:
        return v

    mid = len(v) // 2
    left = merge_sort(v[:mid])
    right = merge_sort(v[mid:])
    return merge(left, right)

print(merge_sort([38, 27, 43, 3, 9, 82, 10]))
````

## Complexitate
- timp: O(n log n) in toate cazurile;
- memorie: O(n).

## Avantaje
- rapid;
- stabil;
- comportament predictibil;
- foarte bun pentru liste mari.

## Dezavantaje
- foloseste memorie suplimentara;
- mai complex decat algoritmii elementari.

## Exercitiu rezolvat
Numarul de combinari:


````python
def merge_with_count(left, right):
    result = []
    i = 0
    j = 0
    comparisons = 0

    while i < len(left) and j < len(right):
        comparisons += 1
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result, comparisons

def merge_sort_with_count(v):
    if len(v) <= 1:
        return v, 0

    mid = len(v) // 2
    left, c1 = merge_sort_with_count(v[:mid])
    right, c2 = merge_sort_with_count(v[mid:])
    merged, c3 = merge_with_count(left, right)

    return merged, c1 + c2 + c3

print(merge_sort_with_count([38, 27, 43, 3, 9, 82, 10]))
````

# 7. Quick Sort

## Idee
Quick Sort alege un pivot si imparte elementele in:
- mai mici decat pivotul;
- egale cu pivotul;
- mai mari decat pivotul.

Apoi sorteaza recursiv partile din stanga si din dreapta.

## Exemplu pas cu pas
Lista:
```text
[10, 7, 8, 9, 1, 5]
```

Alegem pivotul 8:
- stanga -> [7, 1, 5]
- mijloc -> [8]
- dreapta -> [10, 9]

Sortam recursiv:
- [7, 1, 5] -> [1, 5, 7]
- [10, 9] -> [9, 10]

Rezultatul final:
```text
[1, 5, 7, 8, 9, 10]
```


````python
def quick_sort(v):
    if len(v) <= 1:
        return v

    pivot = v[len(v) // 2]
    left = [x for x in v if x < pivot]
    middle = [x for x in v if x == pivot]
    right = [x for x in v if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

print(quick_sort([10, 7, 8, 9, 1, 5]))
````

## Variante utile
### Pivot = primul element


````python
def quick_sort_first_pivot(v):
    if len(v) <= 1:
        return v

    pivot = v[0]
    left = [x for x in v[1:] if x < pivot]
    middle = [pivot] + [x for x in v[1:] if x == pivot]
    right = [x for x in v[1:] if x > pivot]

    return quick_sort_first_pivot(left) + middle + quick_sort_first_pivot(right)

print(quick_sort_first_pivot([10, 7, 8, 9, 1, 5]))
````

### Varianta in-place


````python
def partition(v, low, high):
    pivot = v[high]
    i = low - 1

    for j in range(low, high):
        if v[j] <= pivot:
            i += 1
            v[i], v[j] = v[j], v[i]

    v[i + 1], v[high] = v[high], v[i + 1]
    return i + 1

def quick_sort_inplace(v, low, high):
    if low < high:
        pi = partition(v, low, high)
        quick_sort_inplace(v, low, pi - 1)
        quick_sort_inplace(v, pi + 1, high)

v = [10, 7, 8, 9, 1, 5]
quick_sort_inplace(v, 0, len(v) - 1)
print(v)
````

## Complexitate
- caz mediu: O(n log n);
- caz nefavorabil: O(n²);
- memorie: depinde de implementare.

## Avantaje
- foarte rapid in practica;
- foarte important conceptual;
- bun pentru uz general.

## Dezavantaje
- pivot prost ales poate strica performanta;
- nu este stabil in forma clasica;
- varianta in-place este mai greu de inteles.

## Recomandare
Este excelent pentru performanta buna in practica, mai ales daca alegerea pivotului este bine facuta.


# 8. Radix Sort

## Idee
Radix Sort sorteaza numerele cifra cu cifra:
- unitati;
- zeci;
- sute;
- etc.

Nu compara direct toate elementele intre ele, ci foloseste o sortare auxiliara stabila pentru fiecare pozitie.

## Exemplu pas cu pas
Lista:
```text
[170, 45, 75, 90, 802, 24, 2, 66]
```

Mai intai sortam dupa unitati, apoi dupa zeci, apoi dupa sute.
Dupa procesarea tuturor pozitiilor, lista devine complet sortata.


````python
def counting_sort(v, exp):
    n = len(v)
    output = [0] * n
    count = [0] * 10

    for i in range(n):
        index = v[i] // exp
        count[index % 10] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    i = n - 1
    while i >= 0:
        index = v[i] // exp
        output[count[index % 10] - 1] = v[i]
        count[index % 10] -= 1
        i -= 1

    for i in range(n):
        v[i] = output[i]

def radix_sort(v):
    if not v:
        return v

    max1 = max(v)
    exp = 1
    while max1 // exp > 0:
        counting_sort(v, exp)
        exp *= 10
    return v

print(radix_sort([170, 45, 75, 90, 802, 24, 2, 66]))
````

## Complexitate
- timp: O(nk), unde k este numarul de cifre;
- memorie: O(n + baza).

## Avantaje
- foarte rapid pentru numere intregi potrivite;
- poate depasi algoritmii bazati pe comparatii in anumite situatii.

## Dezavantaje
- nu este universal;
- necesita adaptari pentru unele tipuri de date;
- mai greu de inteles la inceput.

## Recomandare
Foarte bun pentru numere intregi nenegative si chei de forma controlata.


# 9. Comparatie experimentala intre algoritmi

In aceasta sectiune masuram experimental timpii de executie pentru:
- Bubble Sort
- Selection Sort
- Insertion Sort
- Merge Sort
- Quick Sort
- Radix Sort


````python
import random
import time

def time_function(func, data, repeats=3):
    best = float("inf")
    for _ in range(repeats):
        test_data = data[:]
        start = time.perf_counter()
        func(test_data)
        elapsed = time.perf_counter() - start
        best = min(best, elapsed)
    return best

def merge_sort_wrapper(v):
    return merge_sort(v)

def quick_sort_wrapper(v):
    return quick_sort(v)

def radix_sort_wrapper(v):
    radix_sort(v)
    return v

sizes = [100, 500, 1000]
results = []

for n in sizes:
    data = [random.randint(0, 10000) for _ in range(n)]
    algorithms = [
        ("Bubble Sort", bubble_sort),
        ("Selection Sort", selection_sort),
        ("Insertion Sort", insertion_sort),
        ("Merge Sort", merge_sort_wrapper),
        ("Quick Sort", quick_sort_wrapper),
        ("Radix Sort", radix_sort_wrapper),
    ]
    for name, func in algorithms:
        t = time_function(func, data)
        results.append((n, name, t))

results
````

## Interpretare
In mod normal:
- algoritmii O(n²) devin rapid lenti pe liste mari;
- Merge Sort si Quick Sort sunt mult mai buni pentru volume mari;
- Radix Sort poate fi excelent pentru numere intregi.


# 10. Vizualizare grafica a timpilor


````python
import matplotlib.pyplot as plt

algorithms = sorted(set(name for _, name, _ in results))

for alg in algorithms:
    xs = []
    ys = []
    for n, name, t in results:
        if name == alg:
            xs.append(n)
            ys.append(t)
    plt.plot(xs, ys, marker="o", label=alg)

plt.xlabel("Numar elemente")
plt.ylabel("Timp (secunde)")
plt.title("Comparatie experimentala intre algoritmi de sortare")
plt.legend()
plt.show()
````

# 14. Concluzii

Nu exista un algoritm perfect pentru toate situatiile.

### De retinut
- Bubble Sort si Selection Sort sunt excelenti pentru intelegerea mecanismelor de sortare de baza.
- Insertion Sort este foarte bun pe liste mici sau aproape sortate.
- Merge Sort ofera performanta stabila si previzibila.
- Quick Sort este foarte rapid in practica.
- Radix Sort este excelent pe tipurile potrivite de date.

Alegerea algoritmului depinde de:
- dimensiunea listei;
- tipul datelor;
- nevoia de stabilitate;
- memoria disponibila;
- performanta dorita.
