## 6. Metoda Brute Force (Forta bruta)

Metoda brute force (forta bruta) consta in:
- generarea tuturor solutiilor posibile;
- verificarea fiecareia;
- alegerea solutiei optime sau valide.

Este cea mai directa abordare si nu necesita optimizari sofisticate.

- nu foloseste euristici;
- nu elimina solutii din start;
- garanteaza gasirea solutiei optime (daca exista);
- de obicei este ineficienta pentru probleme mari.


## Exemplu: cautarea unui element

Verificam fiecare element din lista pana il gasim.


````python
def linear_search(v, x):
    for i in range(len(v)):
        if v[i] == x:
            return i
    return -1

print(linear_search([4, 2, 7, 1, 9], 7))
````

### Complexitate
```text
O(n)
```
Trebuie sa verificam fiecare element in cel mai rau caz.


## Exemplu: generarea tuturor submultimilor

Pentru o multime de n elemente exista:
```text
2^n
```
submultimi.


````python
def subsets(v):
    n = len(v)
    for mask in range(1 << n):
        subset = []
        for i in range(n):
            if mask & (1 << i):
                subset.append(v[i])
        print(subset)

subsets([1,2,3])
````

### Complexitate
```text
O(2^n)
```
Numarul de solutii creste exponential.


## Exemplu: problema rucsacului (0/1) brute force

Incercam toate combinatiile posibile de obiecte.


````python
def knapsack_bruteforce(weights, values, capacity):
    n = len(weights)
    best = 0

    for mask in range(1 << n):
        total_w = 0
        total_v = 0

        for i in range(n):
            if mask & (1 << i):
                total_w += weights[i]
                total_v += values[i]

        if total_w <= capacity:
            best = max(best, total_v)

    return best

weights = [10,20,30]
values = [60,100,120]
print(knapsack_bruteforce(weights, values, 50))
````

### Complexitate
```text
O(2^n)
```

Aceasta explica de ce metoda brute force devine rapid impracticabila pentru n mare.


## Avantaje si limite

### Avantaje
- simplu de implementat;
- garanteaza solutia optima;
- bun pentru probleme mici sau verificare.

### Dezavantaje
- foarte lent pentru date mari;
- complexitate exponentiala sau factoriala;
- nu scaleaza.

## Rolul metodei brute force

Este foarte importanta deoarece:
- reprezinta punctul de plecare pentru optimizari;
- ajuta la verificarea corectitudinii altor algoritmi;
- este baza pentru metode mai avansate (backtracking, DP).


# Metode fundamentale de proiectare a algoritmilor

## Cuprins
7. **Greedy**: alegerea locala optima;
8. **Divide et Impera**: descompunerea recursiva a problemei;
9. **Backtracking**: explorarea sistematica a spatiului solutiilor.




## Legatura dintre Brute Force, Greedy, Divide et Impera si Backtracking

Aceste metode sunt strans legate, dar diferă fundamental prin modul în care abordeaza o problema.

---

### Brute Force
Brute Force incearca toate solutiile posibile.  
Nu elimina nimic din start.  
Este cea mai directa si generala abordare.

---

### Backtracking
Backtracking porneste de la ideea de brute force, dar adauga un mecanism esential:
- daca o solutie partiala nu mai poate duce la o solutie valida, explorarea acelei ramuri este oprita imediat.

Prin urmare:
- brute force exploreaza tot;
- backtracking exploreaza inteligent si taie ramurile imposibile.

---

### Divide et Impera
Divide et Impera nu exploreaza toate solutiile, ci:
- imparte problema in subprobleme mai mici;
- rezolva recursiv fiecare subproblema;
- combina rezultatele pentru a obtine solutia finala.

Este eficient atunci cand:
- problema poate fi descompusa in subprobleme independente;
- combinarea solutiilor este eficienta.

Prin urmare:
- nu cauta in spatiul solutiilor;
- nu elimina variante, ci restructureaza problema.

---

### Greedy
Greedy nu exploreaza toate solutiile si nu descompune problema complet recursiv.

In schimb:
- face la fiecare pas o alegere local optima;
- nu revine asupra deciziilor;
- construieste solutia incremental.

Este corect doar daca:
- alegerea locala conduce la optimul global.

Prin urmare:
- este foarte rapid;
- dar are aplicabilitate limitata.

---

## Sinteza

- **Brute Force**: exploreaza toate solutiile posibile  
- **Backtracking**: brute force + eliminarea ramurilor imposibile  
- **Divide et Impera**: descompune problema recursiv  
- **Greedy**: face alegeri locale rapide fara explorare completa  

---

## Relatia dintre metode

Se poate privi progresiv astfel:

1. Pornim de la **Brute Force** (incercam tot)
2. Eliminam ramurile imposibile -> **Backtracking**
3. Descompunem problema in subprobleme -> **Divide et Impera**
4. Facem alegeri locale sigure -> **Greedy**

---

Aceste metode sunt complementare, nu concurente.

Alegerea metodei depinde de:
- structura problemei;
- constrangeri;
- dimensiunea datelor;
- tipul rezultatului dorit (toate solutiile vs optim).


## Exemplu intuitiv: Fibonacci

### Varianta recursiva simpla
Pentru a calcula `F(n)`:
```text
F(n) = F(n-1) + F(n-2)
```

Aceasta formula produce multe recalculari.
De exemplu, `F(3)` poate fi calculat de mai multe ori in aceeasi executie.

### Interpretare
- o varianta recursiva naiva seamana cu o abordare brute force asupra arborelui de apeluri;

Acest exemplu este foarte util pentru a intelege de ce simpla cautare exhaustiva poate fi ineficienta si cum poate fi imbunatatita.


````python
def fib_naiv(n):
    if n <= 1:
        return n
    return fib_naiv(n - 1) + fib_naiv(n - 2)

def fib_dp(n):
    if n <= 1:
        return n

    dp = [0] * (n + 1)
    dp[1] = 1

    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]

    return dp[n]

print(fib_naiv(10))
print(fib_dp(10))
````

## Cum aleg metoda potrivita?

### 1. Pot incerca toate solutiile?
Daca da, **Brute Force** este un punct de plecare bun.  
Este simplu si garanteaza solutia corecta, dar poate deveni rapid ineficient.

---

### 2. Pot elimina devreme solutii partiale invalide?
Daca da, **Backtracking** este o imbunatatire naturala.  
Reducem spatiul de cautare eliminand ramurile care nu pot duce la solutii valide.

---

### 3. Pot descompune problema in subprobleme mai mici?
Daca da, **Divide et Impera** este potrivit.  
Problema este impartita recursiv in bucati mai simple, rezolvate independent si apoi combinate.

---

### 4. Exista o alegere locala sigura?
Daca da, poate exista o solutie **Greedy**.  
Algoritmul construieste solutia pas cu pas, fara sa exploreze toate posibilitatile.

---

## Ideea-cheie

Aceasta progresie este foarte utila in proiectarea algoritmilor:

1. Incepem cu **Brute Force**  
2. Eliminam ramurile inutile -> **Backtracking**  
3. Reorganizăm problema -> **Divide et Impera**  
4. Simplificam decizia -> **Greedy**

Intelegerea acestui proces face mult mai usoara alegerea metodei potrivite pentru o problema noua.


## Rezumat

- **Brute Force** este abordarea de baza: incercam toate solutiile posibile.
- **Backtracking** este brute force imbunatatit: eliminam din timp ramurile care nu pot duce la solutii valide.
- **Divide et Impera** rezolva problema prin descompunerea in subprobleme mai mici, rezolvate recursiv si apoi combinate.
- **Greedy** urmeaza o alta strategie: nu exploreaza toate posibilitatile, ci face alegeri locale rapide, sperand ca acestea conduc la optimul global.


# 7. Metoda Greedy

## 7.1 Ideea de baza

Metoda Greedy construieste o solutie pas cu pas.  
La fiecare etapa, algoritmul alege optiunea care pare cea mai buna **in acel moment**.

Aceasta alegere este:
- locala;
- imediata;
- de regula ireversibila.

Algoritmul nu exploreaza toate posibilitatile si nu revine asupra alegerilor facute.  
Din acest motiv, este de obicei foarte eficient.

Totusi, metoda Greedy nu functioneaza pentru orice problema.  
Pentru a fi corecta, problema trebuie sa aiba anumite proprietati structurale.


## 7.2 Optim local versus optim global

Aceasta distinctie este esentiala.

### Optim local
Cea mai buna alegere la pasul curent, fara a analiza toate consecintele viitoare.

### Optim global
Cea mai buna solutie dintre toate solutiile posibile.

O metoda Greedy este corecta doar daca alegerea optima locala conduce, in final, la o solutie optima globala.

Aceasta nu este o proprietate automata.  
Ea trebuie demonstrata sau justificata pentru fiecare problema in parte.


## 7.3 Proprietati teoretice importante

Pentru ca o problema sa poata fi rezolvata prin Greedy, apar frecvent doua proprietati:

### 1. Substructura optima
O solutie optima a problemei poate fi construita din solutii optime ale unor subprobleme.

### 2. Proprietatea de alegere Greedy
Exista o alegere locala optima care poate face parte dintr-o solutie optima globala.

Aceasta a doua proprietate este cea mai specifica metodei Greedy.  
Daca nu poate fi justificata, atunci algoritmul Greedy poate produce o solutie suboptima.


## 7.4 Schema tipica de demonstratie a corectitudinii

O demonstratie clasica foloseste un **argument de schimb**.

Ideea este urmatoarea:
1. Presupunem ca exista o solutie optima.
2. Aratam ca, daca aceasta nu contine alegerea Greedy, o putem transforma intr-o alta solutie optima care o contine.
3. Prin urmare, alegerea Greedy este sigura.

Acest tip de argument este foarte important in analiza algoritmilor Greedy.


## 7.5 Exemplu rezolvat: selectia activitatilor

### Enunt
Avem un set de activitati, fiecare descrisa prin:
- moment de inceput;
- moment de sfarsit.

Dorim sa alegem un numar maxim de activitati compatibile, adica activitati care nu se suprapun.

### Strategie Greedy
Sortam activitatile dupa momentul de finalizare si alegem mereu activitatea care se termina cel mai devreme dintre cele compatibile cu cele deja selectate.


````python
def activity_selection(activities):
    activities = sorted(activities, key=lambda x: x[1])
    selected = []
    last_end = 0

    for start, end in activities:
        if start >= last_end:
            selected.append((start, end))
            last_end = end

    return selected


activities = [
    (1, 4), (3, 5), (0, 6), (5, 7), (3, 9),
    (5, 9), (6, 10), (8, 11), (8, 12),
    (2, 14), (12, 16)
]

print(activity_selection(activities))
````

### Explicatie pas cu pas

Dupa sortarea dupa timpul de finalizare, activitatile sunt analizate in ordinea in care se termina.

Se alege:
- activitatea care se termina cel mai devreme;
- apoi urmatoarea activitate care incepe dupa terminarea celei alese;
- si asa mai departe.

### De ce functioneaza?
Alegerea activitatii care se termina cel mai devreme lasa cat mai mult loc pentru activitatile viitoare.

Aceasta este cheia corectitudinii:
alegerea locala nu blocheaza posibilitatea de a alege multe activitati ulterior.


### Complexitate
- sortarea: `O(n log n)`
- parcurgerea liniara: `O(n)`

Complexitatea totala:
```text
O(n log n)
```


## 7.6 Exemplu rezolvat: problema monetara

### Enunt
Avem monede de valori date si vrem sa obtinem o suma folosind un numar minim de monede.

### Strategie Greedy
La fiecare pas alegem cea mai mare moneda care nu depaseste suma ramasa.


````python
def greedy_coin_change(sum_value, coins):
    coins = sorted(coins, reverse=True)
    result = []

    for coin in coins:
        while sum_value >= coin:
            result.append(coin)
            sum_value -= coin

    return result


coins = [1, 5, 10, 50]
print(greedy_coin_change(93, coins))
````

### Explicatie
Pentru suma 93:
- alegem 50 -> ramane 43
- alegem 10 -> ramane 33
- alegem 10 -> ramane 23
- alegem 10 -> ramane 13
- alegem 10 -> ramane 3
- alegem 1, 1, 1

### Observatie foarte importanta
Aceasta strategie functioneaza pentru multe sisteme monetare uzuale, dar **nu pentru toate**.

Exemplu de contraexemplu:
monede = `[1, 3, 4]`, suma = `6`

Greedy alege:
- 4 + 1 + 1 -> 3 monede

Dar optimul este:
- 3 + 3 -> 2 monede

Deci metoda Greedy nu este universal corecta pentru problema monetara generala.


## 7.7 Exemplu rezolvat: Fractional Knapsack

### Enunt
Avem obiecte cu:
- valoare;
- greutate.

Putem lua si **fractii** din obiecte.  
Dorim sa maximizam valoarea obtinuta intr-un rucsac cu capacitate limitata.

### Strategie Greedy
Sortam obiectele dupa raportul:
```text
valoare / greutate
```
in ordine descrescatoare si luam mai intai obiectele cu densitatea valorica cea mai mare.


````python
def fractional_knapsack(items, capacity):
    # items = (value, weight)
    items = sorted(items, key=lambda x: x[0] / x[1], reverse=True)

    total_value = 0
    taken = []

    for value, weight in items:
        if capacity == 0:
            break

        if weight <= capacity:
            taken.append((value, weight, 1.0))
            total_value += value
            capacity -= weight
        else:
            fraction = capacity / weight
            taken.append((value, weight, fraction))
            total_value += value * fraction
            capacity = 0

    return total_value, taken


items = [(60, 10), (100, 20), (120, 30)]
print(fractional_knapsack(items, 50))
````

### Observatie
Pentru varianta fractionara, strategia Greedy este corecta.

Pentru varianta 0/1 Knapsack, in care obiectele pot fi luate doar integral sau deloc, aceeasi strategie nu mai este in general corecta.


## 7.8 Avantaje si limite ale metodei Greedy

### Avantaje
- simpla;
- rapida;
- usor de implementat;
- de multe ori foarte eleganta.

### Limite
- nu functioneaza pentru toate problemele de optimizare;
- necesita justificare matematica sau structurala;
- o alegere locala buna poate rata optimul global.

### Cand este potrivita?
- cand exista o regula locala sigura;
- cand problema are proprietatea de alegere Greedy;
- cand performanta este foarte importanta.

# 8. Metoda Divide et Impera

## 8.1 Ideea de baza

Strategia Divide et Impera rezolva o problema prin trei pasi:
1. **Divide** – impartim problema in subprobleme mai mici;
2. **Impera** – rezolvam recursiv subproblemele;
3. **Combina** – reunim solutiile partiale.

Aceasta este una dintre cele mai importante paradigme algoritmice.


## 8.2 Model general de recurenta

Multe algoritmi Divide et Impera conduc la relatii de forma:

```text
T(n) = aT(n/b) + f(n)
```

unde:
- `a` este numarul de subprobleme;
- `n/b` este dimensiunea fiecarei subprobleme;
- `f(n)` este costul impartirii si al combinarii.

Aceasta relatie este esentiala pentru analiza complexitatii.


## 8.3 Teorema Master

Teorema Master este una dintre cele mai importante unelte folosite pentru a estima complexitatea algoritmilor recursivi de tip Divide et Impera.

Se aplica recurentelor de forma:

```text
T(n) = aT(n/b) + f(n)
```

unde:
- `a` = numarul de subprobleme generate;
- `b` = factorul de reducere a dimensiunii;
- `f(n)` = costul impartirii si al combinarii.

Exemple tipice:
- `T(n) = 2T(n/2) + n`
- `T(n) = T(n/2) + 1`
- `T(n) = 4T(n/2) + n`


## 8.4 De ce comparam cu `n^(log_b a)`?

Aceasta este intrebarea centrala.

Pentru recursivitatea:
```text
T(n) = aT(n/b) + f(n)
```

arborele de recursie are:
- la nivelul 0: 1 problema de dimensiune `n`
- la nivelul 1: `a` subprobleme de dimensiune `n/b`
- la nivelul 2: `a^2` subprobleme de dimensiune `n/b^2`
- ...
- la nivelul `k`: `a^k` subprobleme de dimensiune `n/b^k`

Recursia se opreste aproximativ cand:
```text
n / b^k = 1
```

de unde rezulta:
```text
k = log_b n
```

Numarul total de frunze este atunci:
```text
a^(log_b n)
```

Folosind proprietatile logaritmilor:
```text
a^(log_b n) = n^(log_b a)
```

Aceasta expresie descrie marimea contributiei recursiei pure.  
De aceea `n^(log_b a)` este termenul cu care comparam `f(n)`.


## 8.5 Intuitia arborelui de recursie

Privim executia recursiva drept un arbore.

La nivelul `i`:
- exista `a^i` subprobleme;
- fiecare are dimensiunea `n / b^i`.

Costul total pe nivelul `i` este:
```text
a^i * f(n / b^i)
```

Acum apar trei posibilitati:

1. costurile cresc spre frunze  
2. costurile sunt aproximativ egale pe fiecare nivel  
3. costurile sunt cele mai mari la radacina

Aceste trei situatii corespund exact celor trei cazuri din Teorema Master.


## 8.6 Cazul 1 – recursia domina

Daca:
```text
f(n) = O(n^(log_b a - ε))
```
pentru un `ε > 0`, atunci:
```text
T(n) = Θ(n^(log_b a))
```

### Intuitie
Costul combinarii este relativ mic comparativ cu volumul de munca generat de subprobleme.

Cu alte cuvinte:
- partea recursiva produce costul dominant;
- frunzele si nivelurile de jos domina totalul.

### Exemplu
```text
T(n) = 4T(n/2) + n
```

Aici:
- `a = 4`
- `b = 2`
- `n^(log_2 4) = n^2`
- `f(n) = n`

Comparam:
```text
n  <  n^2
```

Deci suntem in cazul 1:
```text
T(n) = Θ(n^2)
```


## 8.7 Cazul 2 – echilibru

Daca:
```text
f(n) = Θ(n^(log_b a))
```
atunci:
```text
T(n) = Θ(n^(log_b a) log n)
```

### Intuitie
Costul combinarii si costul total al recursiei sunt de acelasi ordin.

Asta inseamna ca:
- fiecare nivel al arborelui de recursie contribuie aproximativ la fel;
- avem aproximativ `log n` niveluri;
- deci totalul este cost pe nivel × numar de niveluri.

### Exemplu clasic
```text
T(n) = 2T(n/2) + n
```

Aici:
- `a = 2`
- `b = 2`
- `n^(log_2 2) = n`
- `f(n) = n`

Deci:
```text
T(n) = Θ(n log n)
```

Acesta este exact cazul lui Merge Sort.


## 8.8 Cazul 3 – combinarea domina

Daca:
```text
f(n) = Ω(n^(log_b a + ε))
```
pentru un `ε > 0`, si este satisfacuta conditia tehnica standard de regularitate, atunci:
```text
T(n) = Θ(f(n))
```

### Intuitie
Costul impartirii si al combinarii este mult mai mare decat costul total produs de recursie.

Cu alte cuvinte:
- nivelul superior al arborelui este cel mai scump;
- pe masura ce coboram, contributia scade;
- rezultatul final este dominat de `f(n)`.

### Exemplu
```text
T(n) = 2T(n/2) + n^2
```

Aici:
- `a = 2`
- `b = 2`
- `n^(log_2 2) = n`
- `f(n) = n^2`

Deoarece:
```text
n^2 > n
```

rezulta:
```text
T(n) = Θ(n^2)
```


## 8.9 Rezumat intuitiv al celor trei cazuri

Putem retine foarte simplu astfel:

### Cazul 1
Recursia domina.  
Costul total este concentrat la frunze.

### Cazul 2
Recursia si combinarea sunt echilibrate.  
Fiecare nivel contribuie comparabil.  
Apare factorul `log n`.

### Cazul 3
Combinarea domina.  
Costul total este concentrat aproape de radacina.

Aceasta interpretare este adesea mai utila decat memorarea seaca a formulelor.

5. Explicati intuitiv, prin arborele de recursie, de ce la Merge Sort apare factorul `log n`.


## 8.3 Intuitie pentru Teorema Master

Teorema Master ofera o regula rapida pentru estimarea complexitatii recurentelor de forma:
```text
T(n) = aT(n/b) + f(n)
```

Se compara functia `f(n)` cu:
```text
n^(log_b a)
```

### Intuitie
- daca partea recursiva produce costul dominant, recursia domina;
- daca partea de combinare este mai grea, combinarea domina;
- daca cele doua sunt echilibrate, apare adesea un factor `log n`.


## 8.4 Cazurile clasice ale Teoremei Master

### Cazul 1
Daca:
```text
f(n) = O(n^(log_b a - ε))
```
pentru un `ε > 0`, atunci:
```text
T(n) = Θ(n^(log_b a))
```

### Cazul 2
Daca:
```text
f(n) = Θ(n^(log_b a))
```
atunci:
```text
T(n) = Θ(n^(log_b a) log n)
```

### Cazul 3
Daca:
```text
f(n) = Ω(n^(log_b a + ε))
```
si sunt indeplinite conditiile tehnice standard, atunci:
```text
T(n) = Θ(f(n))
```


## 8.5 Exemplu rezolvat: cautarea binara

### Enunt
Intr-un vector sortat, cautam o valoare data.

### Idee
Comparam valoarea cautata cu elementul din mijloc:
- daca este egal, am gasit raspunsul;
- daca este mai mica, cautam in jumatatea stanga;
- daca este mai mare, cautam in jumatatea dreapta.


````python
def binary_search(v, x, left, right):
    if left > right:
        return -1

    mid = (left + right) // 2

    if v[mid] == x:
        return mid
    elif x < v[mid]:
        return binary_search(v, x, left, mid - 1)
    else:
        return binary_search(v, x, mid + 1, right)


v = [2, 5, 7, 10, 14, 18, 21]
print(binary_search(v, 14, 0, len(v) - 1))
print(binary_search(v, 6, 0, len(v) - 1))
````

### Analiza complexitatii
Recursivitatea este:
```text
T(n) = T(n/2) + O(1)
```

Aplicand Teorema Master sau rationamentul direct:
```text
T(n) = O(log n)
```


## 8.6 Exemplu rezolvat: Merge Sort

### Idee
Impartim vectorul in doua jumatati:
- sortam recursiv jumatatea stanga;
- sortam recursiv jumatatea dreapta;
- combinam cele doua liste sortate.


````python
def merge(left, right):
    result = []
    i = 0
    j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


def merge_sort(v):
    if len(v) <= 1:
        return v

    mid = len(v) // 2
    left = merge_sort(v[:mid])
    right = merge_sort(v[mid:])

    return merge(left, right)


print(merge_sort([38, 27, 43, 3, 9, 82, 10]))
````

### Analiza complexitatii
Recursivitatea este:
```text
T(n) = 2T(n/2) + O(n)
```

Aici:
- `a = 2`
- `b = 2`
- `n^(log_b a) = n`

Suntem in cazul 2 al Teoremei Master:
```text
T(n) = Θ(n log n)
```


## 8.7 Exemplu rezolvat: maximul dintr-un vector

### Idee
Impartim vectorul in doua:
- gasim maximul in stanga;
- gasim maximul in dreapta;
- alegem maximul dintre cele doua.


````python
def max_divide(v, left, right):
    if left == right:
        return v[left]

    mid = (left + right) // 2
    max_left = max_divide(v, left, mid)
    max_right = max_divide(v, mid + 1, right)

    return max(max_left, max_right)


v = [7, 2, 18, 4, 15, 1]
print(max_divide(v, 0, len(v) - 1))
````

### Observatie
Aceasta problema poate fi rezolvata si iterativ foarte simplu.  
Totusi, exemplul este util pentru a intelege cum o problema poate fi descompusa recursiv in doua subprobleme similare.


## 8.8 Arborele de recursie – intuitie

Pentru multe recurente Divide et Impera, este util sa privim executia ca pe un arbore:
- fiecare nod reprezinta o subproblema;
- fiecare nivel reprezinta un pas de descompunere;
- costul total este suma costurilor de pe toate nivelurile.

Exemplu pentru Merge Sort:
- pe fiecare nivel se proceseaza in total `n` elemente;
- exista aproximativ `log n` niveluri;
- deci costul total este `n log n`.

### Limite

# 9. Metoda Backtracking

## 9.1 Ideea de baza

Backtracking este o metoda de cautare sistematica in spatiul solutiilor.

Construim o solutie pas cu pas:
- alegem o valoare;
- verificam daca solutia partiala ramane valida;
- daca da, continuam;
- daca nu, revenim si incercam alta alegere.

Aceasta revenire este esenta metodei.


## 9.2 Forta bruta si spatiul solutiilor

Multe probleme pot fi privite astfel:
- exista un numar mare de configuratii posibile;
- vrem sa gasim una sau toate configuratiile valide.

Forta bruta inseamna sa incercam tot.

Backtracking este o imbunatatire:
- nu mai exploram complet ramurile care sunt deja imposibile sau invalide;
- facem **pruning**, adica taiem ramurile nepromitatoare.


## 9.3 Model conceptual

Putem vedea backtracking-ul ca pe un arbore al deciziilor:
- fiecare nivel corespunde unei noi alegeri;
- fiecare nod reprezinta o solutie partiala;
- frunzele reprezinta solutii complete sau ramuri respinse.

Algoritmul merge in adancime:
- exploreaza o ramura;
- revine;
- exploreaza alta ramura.


## 9.4 Exemplu rezolvat: toate sirurile binare de lungime n

### Enunt
Sa se genereze toate sirurile de 0 si 1 de lungime `n`.

### Idee
La fiecare pozitie avem doua alegeri:
- 0
- 1


````python
def generate_binary(n, solution=None):
    if solution is None:
        solution = []

    if len(solution) == n:
        print(solution)
        return

    for value in [0, 1]:
        solution.append(value)
        generate_binary(n, solution)
        solution.pop()


generate_binary(3)
````

### Observatie
Aici nu exista conditii de invalidare, deci practic exploram complet arborele binar al tuturor posibilitatilor.


## 9.5 Exemplu rezolvat: permutari

### Enunt
Sa se genereze toate permutarile numerelor `1..n`.

### Idee
Construim permutarea pas cu pas.
La fiecare pas alegem un numar care nu a fost folosit deja.


````python
def permutations_backtracking(n, solution=None, used=None):
    if solution is None:
        solution = []
    if used is None:
        used = [False] * (n + 1)

    if len(solution) == n:
        print(solution)
        return

    for value in range(1, n + 1):
        if not used[value]:
            used[value] = True
            solution.append(value)

            permutations_backtracking(n, solution, used)

            solution.pop()
            used[value] = False


permutations_backtracking(3)
````

### Complexitate
Numarul total de permutari este:
```text
n!
```

Prin urmare, complexitatea este in mod natural exponentiala / factoriala.


## 9.6 Exemplu rezolvat: combinatii

### Enunt
Sa se genereze toate combinatiile de `k` elemente din multimea `{1, 2, ..., n}`.

### Idee
Construim solutia in ordine crescatoare, ceea ce evita duplicarea acelorasi combinatii in ordine diferita.


````python
def combinations_backtracking(n, k, start=1, solution=None):
    if solution is None:
        solution = []

    if len(solution) == k:
        print(solution)
        return

    for value in range(start, n + 1):
        solution.append(value)
        combinations_backtracking(n, k, value + 1, solution)
        solution.pop()


combinations_backtracking(5, 3)
````

## 9.7 Exemplu rezolvat: problema celor n regine

### Enunt
Trebuie sa plasam `n` regine pe o tabla `n x n` astfel incat:
- sa nu fie doua pe aceeasi coloana;
- sa nu fie doua pe aceeasi diagonala.

### Idee
Punem cate o regina pe fiecare linie.
Pentru fiecare linie incercam toate coloanele valide.


````python
def solve_n_queens(n):
    solutions = []
    columns = set()
    diag1 = set()   # row - col
    diag2 = set()   # row + col
    board = [-1] * n

    def backtrack(row):
        if row == n:
            solutions.append(board[:])
            return

        for col in range(n):
            if col in columns or (row - col) in diag1 or (row + col) in diag2:
                continue

            board[row] = col
            columns.add(col)
            diag1.add(row - col)
            diag2.add(row + col)

            backtrack(row + 1)

            columns.remove(col)
            diag1.remove(row - col)
            diag2.remove(row + col)
            board[row] = -1

    backtrack(0)
    return solutions


print(solve_n_queens(4))
````

### Observatie
Aici backtracking-ul este mult mai eficient decat o forta bruta oarba, deoarece pozitiile invalide sunt eliminate imediat.


## 9.8 Tehnici de optimizare in Backtracking

Backtracking devine mai eficient daca:
- verificarea validitatii este rapida;
- ordinea alegerilor este bine aleasa;
- se taie devreme ramurile imposibile;
- se memoreaza uneori anumite configuratii.

Aceste tehnici reduc dramatic timpul real de executie, chiar daca in cel mai rau caz complexitatea ramane mare.


## 9.9 Avantaje si limite

### Avantaje
- foarte flexibil;
- potrivit pentru probleme cu multe constrangeri;
- poate genera toate solutiile;
- permite taieri inteligente ale spatiului de cautare.

### Limite
- poate avea complexitate exponentiala;
- devine lent pentru instante mari;
- necesita modelare atenta.

# Concluzii finale

## Metoda Greedy
- face alegeri locale;
- este eficienta;
- necesita proprietati speciale pentru a fi corecta.

## Divide et Impera
- descompune problema in subprobleme;
- rezolva recursiv;
- combina rezultatele;
- duce la algoritmi foarte eficienti.

## Backtracking
- exploreaza spatiul solutiilor;
- revine cand o alegere nu mai este valida;
- este foarte puternic pentru probleme cu multe constrangeri.

Aceste trei metode sunt esentiale pentru intelegerea proiectarii algoritmilor si ilustreaza trei stiluri distincte de gandire:
- alegere locala;
- descompunere recursiva;
- explorare sistematica cu revenire.
