## 1. Introducere teoretică

### 1.1. Ce este o listă înlănțuită?

O listă înlănțuită este o structură de date liniară alcătuită din **noduri**.  
Fiecare nod conține:
- o **valoare**;
- una sau mai multe **referințe** către alte noduri.

Spre deosebire de vectori sau listele Python clasice, elementele unei liste înlănțuite nu trebuie să fie stocate în memorie în poziții consecutive.

### 1.2. Listă simplu înlănțuită

Într-o listă simplu înlănțuită, fiecare nod conține:
- `data` – informația utilă;
- `next` – referința către nodul următor.

Reprezentare conceptuală:

`[data | next] -> [data | next] -> [data | next] -> None`

Caracteristici:
- parcurgerea se face doar **înainte**;
- inserarea și ștergerea la început sunt foarte eficiente;
- accesul la un element de pe o anumită poziție necesită parcurgere secvențială.

### 1.3. Listă dublu înlănțuită

Într-o listă dublu înlănțuită, fiecare nod conține:
- `data` – informația utilă;
- `next` – referința către nodul următor;
- `prev` – referința către nodul precedent.

Reprezentare conceptuală:

`None <- [prev | data | next] <-> [prev | data | next] <-> [prev | data | next] -> None`

Caracteristici:
- parcurgerea se poate face **în ambele sensuri**;
- ștergerea unui nod cunoscut este mai comodă decât în lista simplă;
- necesită memorie suplimentară pentru referința `prev`.

### 1.4. Avantaje și dezavantaje

#### Avantaje ale listelor înlănțuite
- dimensiune dinamică;
- inserări și ștergeri eficiente, mai ales la capete;
- nu necesită memorie contiguă.

#### Dezavantaje
- acces lent la elementul de pe poziția `i`;
- consum suplimentar de memorie pentru referințe;
- implementare mai complexă decât pentru un tablou/vector.

### 1.5. Când se folosesc?

Listele înlănțuite sunt utile atunci când:
- avem multe inserări și ștergeri;
- dimensiunea colecției se schimbă frecvent;
- vrem să construim alte structuri de date, cum ar fi:
  - stive;
  - cozi;
  - deques;
  - liste de adiacență în grafuri;
  - mecanisme de tip undo/redo;
  - cache-uri LRU.


## 2. Implementarea unei liste simplu înlănțuite

Vom implementa două clase:
- `NodeSLL` – un nod din lista simplu înlănțuită;
- `SinglyLinkedList` – lista propriu-zisă.


```python
# ============================================================
# Implementarea unei liste simplu înlănțuite
# ============================================================

class NodeSLL:
    """Reprezintă un nod dintr-o listă simplu înlănțuită."""

    def __init__(self, data):
        self.data = data      # valoarea stocată în nod
        self.next = None      # referință către nodul următor


class SinglyLinkedList:
    """Implementarea unei liste simplu înlănțuite."""

    def __init__(self):
        self.head = None      # primul nod al listei
        self.size = 0         # numărul de elemente din listă

    def is_empty(self):
        """Verifică dacă lista este goală."""
        return self.head is None

    def append(self, data):
        """Adaugă un element la sfârșitul listei."""
        new_node = NodeSLL(data)

        # Dacă lista este goală, noul nod devine capul listei
        if self.head is None:
            self.head = new_node
            self.size += 1
            return

        # Altfel, mergem până la ultimul nod
        current = self.head
        while current.next:
            current = current.next

        current.next = new_node
        self.size += 1

    def prepend(self, data):
        """Adaugă un element la începutul listei."""
        new_node = NodeSLL(data)
        new_node.next = self.head
        self.head = new_node
        self.size += 1

    def insert_after(self, target, data):
        """Inserează un element după prima apariție a valorii target."""
        current = self.head

        while current:
            if current.data == target:
                new_node = NodeSLL(data)
                new_node.next = current.next
                current.next = new_node
                self.size += 1
                return True
            current = current.next

        return False

    def delete(self, data):
        """Șterge prima apariție a valorii date din listă."""
        current = self.head
        previous = None

        while current:
            if current.data == data:
                # Dacă elementul de șters este primul
                if previous is None:
                    self.head = current.next
                else:
                    previous.next = current.next

                self.size -= 1
                return True

            previous = current
            current = current.next

        return False

    def search(self, data):
        """Caută o valoare în listă și returnează True/False."""
        current = self.head
        while current:
            if current.data == data:
                return True
            current = current.next
        return False

    def to_list(self):
        """Transformă lista înlănțuită într-o listă Python obișnuită."""
        result = []
        current = self.head

        while current:
            result.append(current.data)
            current = current.next

        return result

    def display(self):
        """Afișează lista într-un format ușor de citit."""
        values = []
        current = self.head

        while current:
            values.append(str(current.data))
            current = current.next

        print(" -> ".join(values) + " -> None")

    def __len__(self):
        """Returnează numărul de elemente."""
        return self.size
```

```python
# Exemplu de utilizare pentru lista simplu înlănțuită

sll = SinglyLinkedList()

sll.append(10)
sll.append(20)
sll.append(30)
sll.prepend(5)

print("Lista după inserări:")
sll.display()

print("\nCăutare 20:", sll.search(20))
print("Căutare 99:", sll.search(99))

sll.insert_after(20, 25)
print("\nLista după inserarea lui 25 după 20:")
sll.display()

sll.delete(10)
print("\nLista după ștergerea lui 10:")
sll.display()

print("\nLista ca listă Python:", sll.to_list())
print("Număr de elemente:", len(sll))
```

## 3. Implementarea unei liste dublu înlănțuite

Vom implementa două clase:
- `NodeDLL` – un nod din lista dublu înlănțuită;
- `DoublyLinkedList` – lista propriu-zisă.


```python
# ============================================================
# Implementarea unei liste dublu înlănțuite
# ============================================================

class NodeDLL:
    """Reprezintă un nod dintr-o listă dublu înlănțuită."""

    def __init__(self, data):
        self.data = data      # valoarea stocată
        self.prev = None      # referință către nodul precedent
        self.next = None      # referință către nodul următor


class DoublyLinkedList:
    """Implementarea unei liste dublu înlănțuite."""

    def __init__(self):
        self.head = None      # primul nod
        self.tail = None      # ultimul nod
        self.size = 0

    def is_empty(self):
        """Verifică dacă lista este goală."""
        return self.head is None

    def append(self, data):
        """Adaugă un nod la sfârșitul listei."""
        new_node = NodeDLL(data)

        if self.head is None:
            self.head = self.tail = new_node
            self.size += 1
            return

        self.tail.next = new_node
        new_node.prev = self.tail
        self.tail = new_node
        self.size += 1

    def prepend(self, data):
        """Adaugă un nod la începutul listei."""
        new_node = NodeDLL(data)

        if self.head is None:
            self.head = self.tail = new_node
            self.size += 1
            return

        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node
        self.size += 1

    def delete(self, data):
        """Șterge prima apariție a valorii date."""
        current = self.head

        while current:
            if current.data == data:
                # Legătura cu nodul precedent
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next

                # Legătura cu nodul următor
                if current.next:
                    current.next.prev = current.prev
                else:
                    self.tail = current.prev

                self.size -= 1
                return True

            current = current.next

        return False

    def search(self, data):
        """Caută o valoare și returnează True/False."""
        current = self.head
        while current:
            if current.data == data:
                return True
            current = current.next
        return False

    def display_forward(self):
        """Afișează lista de la început la sfârșit."""
        values = []
        current = self.head

        while current:
            values.append(str(current.data))
            current = current.next

        print(" <-> ".join(values) + " <-> None")

    def display_backward(self):
        """Afișează lista de la sfârșit la început."""
        values = []
        current = self.tail

        while current:
            values.append(str(current.data))
            current = current.prev

        print(" <-> ".join(values) + " <-> None")

    def to_list(self):
        """Transformă lista într-o listă Python standard."""
        result = []
        current = self.head

        while current:
            result.append(current.data)
            current = current.next

        return result

    def __len__(self):
        """Returnează dimensiunea listei."""
        return self.size
```

```python
# Exemplu de utilizare pentru lista dublu înlănțuită

dll = DoublyLinkedList()

dll.append("A")
dll.append("B")
dll.append("C")
dll.prepend("Start")

print("Parcurgere înainte:")
dll.display_forward()

print("\nParcurgere înapoi:")
dll.display_backward()

print("\nCăutare 'B':", dll.search("B"))
print("Căutare 'X':", dll.search("X"))

dll.delete("B")
print("\nDupă ștergerea lui 'B':")
dll.display_forward()

print("\nLista ca listă Python:", dll.to_list())
print("Număr de elemente:", len(dll))
```

### Exercițiul 1. Implementarea unei stive folosind listă simplu înlănțuită

**Cerință:**  
Implementați o stivă (`Stack`) folosind o listă simplu înlănțuită. Operațiile necesare sunt:
- `push(x)` – adaugă un element;
- `pop()` – scoate elementul din vârf;
- `peek()` – afișează elementul din vârf;
- `is_empty()` – verifică dacă stiva este goală.

**Unde se poate utiliza?**
- evaluarea expresiilor;
- paranteze echilibrate;
- mecanisme de undo.


```python
# ============================================================
# Exercițiul 1: Stivă implementată cu listă simplu înlănțuită
# ============================================================

class LinkedStack:
    """Stivă bazată pe listă simplu înlănțuită."""

    def __init__(self):
        self.top = None
        self.size = 0

    def is_empty(self):
        return self.top is None

    def push(self, data):
        # Inserarea într-o stivă se face eficient la început
        new_node = NodeSLL(data)
        new_node.next = self.top
        self.top = new_node
        self.size += 1

    def pop(self):
        # Dacă stiva este goală, nu putem extrage nimic
        if self.is_empty():
            return None

        value = self.top.data
        self.top = self.top.next
        self.size -= 1
        return value

    def peek(self):
        # Returnează elementul din vârf fără a-l elimina
        if self.is_empty():
            return None
        return self.top.data

    def __len__(self):
        return self.size


# Testare
stack = LinkedStack()
stack.push(100)
stack.push(200)
stack.push(300)

print("Elementul din vârf:", stack.peek())
print("Extragem:", stack.pop())
print("Elementul din vârf după pop:", stack.peek())
print("Dimensiune stivă:", len(stack))
```

### Exercițiul 2. Implementarea unei cozi folosind listă simplu înlănțuită

**Cerință:**  
Implementați o coadă (`Queue`) folosind o listă simplu înlănțuită, cu operațiile:
- `enqueue(x)` – adaugă la final;
- `dequeue()` – extrage din față;
- `front()` – arată primul element.

**Unde se poate utiliza?**
- simularea persoanelor la ghișeu;
- task scheduling;
- buffer de procesare.


```python
# ============================================================
# Exercițiul 2: Coadă implementată cu listă simplu înlănțuită
# ============================================================

class LinkedQueue:
    """Coadă bazată pe listă simplu înlănțuită."""

    def __init__(self):
        self.front_node = None
        self.rear_node = None
        self.size = 0

    def is_empty(self):
        return self.front_node is None

    def enqueue(self, data):
        # Adăugarea se face la sfârșit
        new_node = NodeSLL(data)

        if self.is_empty():
            self.front_node = self.rear_node = new_node
        else:
            self.rear_node.next = new_node
            self.rear_node = new_node

        self.size += 1

    def dequeue(self):
        # Extragem primul element din coadă
        if self.is_empty():
            return None

        value = self.front_node.data
        self.front_node = self.front_node.next

        # Dacă am golit coada, actualizăm și capătul din spate
        if self.front_node is None:
            self.rear_node = None

        self.size -= 1
        return value

    def front(self):
        if self.is_empty():
            return None
        return self.front_node.data

    def __len__(self):
        return self.size


# Testare
queue = LinkedQueue()
queue.enqueue("Ana")
queue.enqueue("Bogdan")
queue.enqueue("Carmen")

print("Primul din coadă:", queue.front())
print("Iese din coadă:", queue.dequeue())
print("Primul rămas:", queue.front())
print("Dimensiune coadă:", len(queue))
```

**Unde se poate utiliza?**

- prelucrare secvențială a înregistrărilor.


```python
# ============================================================
# Exercițiul 3: Numărarea elementelor pare
# ============================================================

def count_even_values(linked_list):
    """Numără valorile pare dintr-o listă simplu înlănțuită."""
    count = 0
    current = linked_list.head

    while current:
        # Verificăm dacă valoarea este număr întreg și este pară
        if isinstance(current.data, int) and current.data % 2 == 0:
            count += 1
        current = current.next

    return count


# Testare
numbers = SinglyLinkedList()
for x in [3, 8, 11, 14, 20, 21]:
    numbers.append(x)

numbers.display()
print("Număr de valori pare:", count_even_values(numbers))
```

### Exercițiul 4. Inversarea unei liste simplu înlănțuite

**Cerință:**  
Scrieți o funcție care inversează legăturile unei liste simplu înlănțuite.

**Unde se poate utiliza?**
- procesarea în ordine inversă;
- algoritmi care necesită restructurarea legăturilor;
- exercițiu clasic de interviu.


```python
# ============================================================
# Exercițiul 4: Inversarea unei liste simplu înlănțuite
# ============================================================

def reverse_sll(linked_list):
    """Inversează lista simplu înlănțuită in-place."""
    prev = None
    current = linked_list.head

    while current:
        next_node = current.next   # memorăm nodul următor
        current.next = prev        # inversăm legătura curentă
        prev = current             # mutăm prev înainte
        current = next_node        # continuăm cu următorul nod

    linked_list.head = prev


# Testare
lst = SinglyLinkedList()
for x in [1, 2, 3, 4, 5]:
    lst.append(x)

print("Lista inițială:")
lst.display()

reverse_sll(lst)

print("Lista inversată:")
lst.display()
```

### Exercițiul 5. Istoric înainte/înapoi folosind listă dublu înlănțuită

**Cerință:**  
Simulați un istoric simplificat de browser folosind o listă dublu înlănțuită.

Operații:
- vizitarea unei pagini noi;
- mutare înapoi;
- mutare înainte.

**Unde se poate utiliza?**
- browser web;
- galerie foto;
- navigator între ecrane într-o aplicație.


```python
# ============================================================
# Exercițiul 5: Browser history cu listă dublu înlănțuită
# ============================================================

class BrowserHistory:
    """Simulare simplă a istoricului unui browser."""

    def __init__(self, homepage):
        self.current = NodeDLL(homepage)

    def visit(self, url):
        # Când vizităm o pagină nouă, 'rupem' ramura forward
        new_node = NodeDLL(url)
        self.current.next = new_node
        new_node.prev = self.current
        self.current = new_node

    def back(self):
        # Ne mutăm la pagina precedentă dacă există
        if self.current.prev:
            self.current = self.current.prev
        return self.current.data

    def forward(self):
        # Ne mutăm la pagina următoare dacă există
        if self.current.next:
            self.current = self.current.next
        return self.current.data

    def get_current_page(self):
        return self.current.data


# Testare
history = BrowserHistory("home.com")
history.visit("curs-python.ro")
history.visit("structuri-date.ro")

print("Pagina curentă:", history.get_current_page())
print("Înapoi:", history.back())
print("Înapoi:", history.back())
print("Înainte:", history.forward())
```

### Exercițiul 6. Playlist muzical cu navigare înainte și înapoi

**Cerință:**  
Implementați un playlist în care putem trece la melodia următoare sau precedentă.

**Unde se poate utiliza?**
- playere audio;
- liste de prezentare;
- navigare între elemente media.


```python
# ============================================================
# Exercițiul 6: Playlist cu listă dublu înlănțuită
# ============================================================

class Playlist:
    """Playlist simplu bazat pe listă dublu înlănțuită."""

    def __init__(self):
        self.songs = DoublyLinkedList()
        self.current = None

    def add_song(self, title):
        self.songs.append(title)

        # Dacă este prima melodie, ea devine și melodia curentă
        if self.current is None:
            self.current = self.songs.head

    def next_song(self):
        if self.current and self.current.next:
            self.current = self.current.next
        return self.current.data if self.current else None

    def previous_song(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
        return self.current.data if self.current else None

    def current_song(self):
        return self.current.data if self.current else None


# Testare
playlist = Playlist()
playlist.add_song("Song A")
playlist.add_song("Song B")
playlist.add_song("Song C")

print("Melodia curentă:", playlist.current_song())
print("Următoarea melodie:", playlist.next_song())
print("Următoarea melodie:", playlist.next_song())
print("Melodia precedentă:", playlist.previous_song())
```

### Exercițiul 7. Ștergerea tuturor aparițiilor unei valori dintr-o listă simplu înlănțuită

**Cerință:**  
Scrieți o funcție care elimină toate nodurile ce conțin o anumită valoare.

**Unde se poate utiliza?**
- curățarea datelor;
- eliminarea valorilor invalide;
- filtrarea unei structuri secvențiale.


```python
# ============================================================
# Exercițiul 7: Ștergerea tuturor aparițiilor unei valori
# ============================================================

def delete_all_occurrences(linked_list, value):
    """Șterge toate aparițiile unei valori dintr-o listă simplu înlănțuită."""

    # Eliminăm valorile egale cu 'value' aflate la începutul listei
    while linked_list.head and linked_list.head.data == value:
        linked_list.head = linked_list.head.next
        linked_list.size -= 1

    current = linked_list.head

    # Parcurgem lista și eliminăm aparițiile din interior
    while current and current.next:
        if current.next.data == value:
            current.next = current.next.next
            linked_list.size -= 1
        else:
            current = current.next


# Testare
vals = SinglyLinkedList()
for x in [2, 5, 2, 7, 2, 9, 2]:
    vals.append(x)

print("Lista inițială:")
vals.display()

delete_all_occurrences(vals, 2)

print("Lista după eliminarea tuturor valorilor 2:")
vals.display()
```

### Exercițiul 8. Detectarea unui palindrom folosind listă dublu înlănțuită

**Cerință:**  
Verificați dacă o secvență de caractere stocată într-o listă dublu înlănțuită formează un palindrom.

**Unde se poate utiliza?**
- validări de șiruri;
- algoritmi de comparare simetrică;
- exerciții educaționale pentru parcurgere din ambele sensuri.


```python
# ============================================================
# Exercițiul 8: Verificare palindrom cu listă dublu înlănțuită
# ============================================================

def build_dll_from_string(text):
    """Construiește o listă dublu înlănțuită dintr-un șir de caractere."""
    dll = DoublyLinkedList()
    for ch in text:
        dll.append(ch)
    return dll


def is_palindrome_dll(dll):
    """Verifică dacă elementele din listă formează un palindrom."""
    left = dll.head
    right = dll.tail

    while left and right and left != right and left.prev != right:
        if left.data != right.data:
            return False
        left = left.next
        right = right.prev

    return True


# Testare
word1 = build_dll_from_string("radar")
word2 = build_dll_from_string("python")

print("'radar' este palindrom?", is_palindrome_dll(word1))
print("'python' este palindrom?", is_palindrome_dll(word2))
```

Listele simplu și dublu înlănțuite reprezintă structuri de date fundamentale în informatică.

### Idei-cheie:
- lista simplu înlănțuită este mai simplă și mai economică;
- lista dublu înlănțuită este mai flexibilă la parcurgere;
- ambele sunt utile pentru modelarea colecțiilor dinamice;
- în practică, ele stau la baza multor structuri și mecanisme software.
