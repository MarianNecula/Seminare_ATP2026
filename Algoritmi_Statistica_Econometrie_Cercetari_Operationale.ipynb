# 1. Statistici descriptive

## 1.1 Media aritmetica

Media este o masura a tendintei centrale.

```text
media = suma valorilor / numarul observatiilor
```

## Pseudocod

```text
Algoritm Media
Intrare: vector x cu n valori
Iesire: media valorilor

suma <- 0
Pentru fiecare valoare v din x:
    suma <- suma + v
media <- suma / n
Returneaza media
```

## Implementare Python

````python
def media(x):
    suma = 0
    for valoare in x:
        suma += valoare
    return suma / len(x)

x = [12, 15, 18, 21, 25]
print(media(x))
````

## 1.2 Varianta si deviata standard

Varianta masoara dispersia valorilor fata de medie.

```text
var = suma((xi - media)^2) / n
```

## Pseudocod

```text
Algoritm Varianta
Intrare: vector x
Iesire: varianta

m <- Media(x)
suma <- 0
Pentru fiecare valoare v din x:
    suma <- suma + (v - m)^2
var <- suma / n
Returneaza var
```

## Implementare Python

````python
import math

def varianta(x):
    m = media(x)
    suma = 0
    for valoare in x:
        suma += (valoare - m) ** 2
    return suma / len(x)

def deviatie_standard(x):
    return math.sqrt(varianta(x))

salarii = [3000, 3500, 4200, 5000, 8000]
print("Salariu mediu:", media(salarii))
print("Varianta:", varianta(salarii))
print("Deviatia standard:", deviatie_standard(salarii))
````

# 2. Corelatia Pearson

Corelatia Pearson masoara intensitatea relatiei liniare dintre doua variabile.

```text
r = cov(x, y) / (std(x) * std(y))
```

Interpretare: `r` aproape de 1 indica relatie pozitiva puternica, `r` aproape de -1 relatie negativa puternica, iar `r` aproape de 0 relatie liniara slaba.

## Pseudocod

```text
Algoritm CorelatiePearson
Intrare: vectori x si y
Iesire: coeficientul r

mx <- Media(x)
my <- Media(y)
numarator <- 0
den_x <- 0
den_y <- 0

Pentru i de la 0 la n-1:
    numarator <- numarator + (x[i] - mx) * (y[i] - my)
    den_x <- den_x + (x[i] - mx)^2
    den_y <- den_y + (y[i] - my)^2

r <- numarator / sqrt(den_x * den_y)
Returneaza r
```

## Implementare Python

````python
def corelatie_pearson(x, y):
    mx = media(x)
    my = media(y)
    numarator = 0
    den_x = 0
    den_y = 0
    for i in range(len(x)):
        dx = x[i] - mx
        dy = y[i] - my
        numarator += dx * dy
        den_x += dx ** 2
        den_y += dy ** 2
    return numarator / math.sqrt(den_x * den_y)

marketing = [5, 10, 15, 20, 25, 30]
vanzari = [8, 12, 18, 22, 25, 31]
print("Corelatia Pearson:", corelatie_pearson(marketing, vanzari))
````

# 3. Regresia liniara simpla

Model:

```text
y = a + b*x
```

Formule:

```text
b = suma((xi - media_x)(yi - media_y)) / suma((xi - media_x)^2)
a = media_y - b * media_x
```

## Pseudocod

```text
Algoritm RegresieLiniaraSimpla
Intrare: vectori x si y
Iesire: a, b

mx <- Media(x)
my <- Media(y)
num <- 0
den <- 0
Pentru i de la 0 la n-1:
    num <- num + (x[i] - mx) * (y[i] - my)
    den <- den + (x[i] - mx)^2
b <- num / den
a <- my - b * mx
Returneaza a, b
```

## Implementare Python

````python
def regresie_liniara_simpla(x, y):
    mx = media(x)
    my = media(y)
    num = 0
    den = 0
    for i in range(len(x)):
        num += (x[i] - mx) * (y[i] - my)
        den += (x[i] - mx) ** 2
    b = num / den
    a = my - b * mx
    return a, b

def predictie_liniara(x, a, b):
    return a + b * x

publicitate = [1, 2, 3, 4, 5, 6]
vanzari = [3, 4, 6, 8, 9, 11]
a, b = regresie_liniara_simpla(publicitate, vanzari)
print("Model: vanzari =", a, "+", b, "* publicitate")
print("Predictie pentru publicitate=7:", predictie_liniara(7, a, b))
````

# 4. Regresia liniara multipla prin Gradient Descent

Model:

```text
y = b0 + b1*x1 + b2*x2 + ... + bp*xp
```

## Pseudocod

```text
Algoritm RegresieMultiplaGradientDescent
Intrare: matrice X, vector y, rata alpha, numar iteratii
Iesire: coeficienti beta

Adauga o coloana de 1 la X pentru intercept
Initializeaza beta cu 0
Pentru fiecare iteratie:
    gradient <- vector de 0
    Pentru fiecare observatie i:
        y_pred <- produs_scalar(beta, X[i])
        eroare <- y_pred - y[i]
        Pentru fiecare coeficient j:
            gradient[j] <- gradient[j] + eroare * X[i][j]
    Pentru fiecare coeficient j:
        beta[j] <- beta[j] - alpha * gradient[j] / n
Returneaza beta
```

## Implementare Python

````python
def adauga_intercept(X):
    X_nou = []
    for rand in X:
        X_nou.append([1] + rand)
    return X_nou

def produs_scalar(a, b):
    suma = 0
    for i in range(len(a)):
        suma += a[i] * b[i]
    return suma

def regresie_multipla_gd(X, y, alpha=0.01, iteratii=1000):
    X = adauga_intercept(X)
    n = len(X)
    p = len(X[0])
    beta = [0] * p
    for _ in range(iteratii):
        gradient = [0] * p
        for i in range(n):
            y_pred = produs_scalar(beta, X[i])
            eroare = y_pred - y[i]
            for j in range(p):
                gradient[j] += eroare * X[i][j]
        for j in range(p):
            beta[j] -= alpha * gradient[j] / n
    return beta

def predictie_multipla(rand, beta):
    rand = [1] + rand
    return produs_scalar(rand, beta)

X = [[12, 1], [14, 2], [16, 4], [18, 6], [20, 8]]
y = [3000, 3500, 4500, 5500, 6500]
beta = regresie_multipla_gd(X, y, alpha=0.0001, iteratii=20000)
print("Coeficienti:", beta)
print("Predictie:", predictie_multipla([17, 5], beta))
````

# 5. Regresia logistica

Regresia logistica este folosita pentru clasificare binara.

```text
p = 1 / (1 + e^(-z))
z = b0 + b1*x1 + ... + bp*xp
```

## Pseudocod

```text
Algoritm RegresieLogistica
Intrare: X, y, alpha, iteratii
Iesire: coeficienti beta

Adauga intercept
Initializeaza beta cu 0
Pentru fiecare iteratie:
    gradient <- vector de 0
    Pentru fiecare observatie:
        z <- produs_scalar(beta, x_i)
        p <- sigmoid(z)
        eroare <- p - y_i
        Actualizeaza gradientul
    beta <- beta - alpha * gradient / n
Returneaza beta
```

## Implementare Python

````python
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

def regresie_logistica(X, y, alpha=0.1, iteratii=1000):
    X = adauga_intercept(X)
    n = len(X)
    p = len(X[0])
    beta = [0] * p
    for _ in range(iteratii):
        gradient = [0] * p
        for i in range(n):
            z = produs_scalar(beta, X[i])
            prob = sigmoid(z)
            eroare = prob - y[i]
            for j in range(p):
                gradient[j] += eroare * X[i][j]
        for j in range(p):
            beta[j] -= alpha * gradient[j] / n
    return beta

def predictie_logistica(rand, beta, prag=0.5):
    rand = [1] + rand
    prob = sigmoid(produs_scalar(beta, rand))
    return 1 if prob >= prag else 0, prob

X = [[0.1, 0], [0.2, 0], [0.4, 1], [0.6, 2], [0.8, 3]]
y = [0, 0, 0, 1, 1]
beta = regresie_logistica(X, y, alpha=0.5, iteratii=5000)
print("Coeficienti:", beta)
print("Client nou:", predictie_logistica([0.7, 2], beta))
````

# 7. Arbore de decizie simplificat

Un arbore de decizie imparte datele in functie de reguli de tip `x_j <= prag`.

Entropia masoara impuritatea unui set.

```text
H = -p0 log2(p0) - p1 log2(p1)
```

## Pseudocod

```text
Algoritm ArboreDecizie
Intrare: date X, etichete y
Iesire: arbore

Daca toate etichetele sunt identice:
    returneaza frunza cu acea clasa
Pentru fiecare variabila:
    Pentru fiecare prag posibil:
        imparte datele in stanga/dreapta
        calculeaza castigul informational
Alege variabila si pragul cu castig maxim
Construieste recursiv subarborele stang si drept
Returneaza nodul
```

## Implementare Python

````python
def entropie(y):
    n = len(y)
    frecvente = {}
    for clasa in y:
        frecvente[clasa] = frecvente.get(clasa, 0) + 1
    h = 0
    for clasa in frecvente:
        p = frecvente[clasa] / n
        h -= p * math.log2(p)
    return h

def impartire(X, y, coloana, prag):
    X_st, y_st = [], []
    X_dr, y_dr = [], []
    for i in range(len(X)):
        if X[i][coloana] <= prag:
            X_st.append(X[i]); y_st.append(y[i])
        else:
            X_dr.append(X[i]); y_dr.append(y[i])
    return X_st, y_st, X_dr, y_dr

def cea_mai_frecventa(y):
    frecvente = {}
    for clasa in y:
        frecvente[clasa] = frecvente.get(clasa, 0) + 1
    return max(frecvente, key=frecvente.get)

def cel_mai_bun_split(X, y):
    h_initial = entropie(y)
    cel_mai_bun = None
    castig_maxim = 0
    n = len(y)
    p = len(X[0])
    for coloana in range(p):
        praguri = sorted(set(rand[coloana] for rand in X))
        for prag in praguri:
            X_st, y_st, X_dr, y_dr = impartire(X, y, coloana, prag)
            if len(y_st) == 0 or len(y_dr) == 0:
                continue
            h_split = (len(y_st) / n) * entropie(y_st) + (len(y_dr) / n) * entropie(y_dr)
            castig = h_initial - h_split
            if castig > castig_maxim:
                castig_maxim = castig
                cel_mai_bun = (coloana, prag)
    return cel_mai_bun

def construieste_arbore(X, y, adancime=0, adancime_maxima=3):
    if len(set(y)) == 1:
        return {"tip": "frunza", "clasa": y[0]}
    if adancime == adancime_maxima:
        return {"tip": "frunza", "clasa": cea_mai_frecventa(y)}
    split = cel_mai_bun_split(X, y)
    if split is None:
        return {"tip": "frunza", "clasa": cea_mai_frecventa(y)}
    coloana, prag = split
    X_st, y_st, X_dr, y_dr = impartire(X, y, coloana, prag)
    return {"tip": "nod", "coloana": coloana, "prag": prag,
            "stanga": construieste_arbore(X_st, y_st, adancime + 1, adancime_maxima),
            "dreapta": construieste_arbore(X_dr, y_dr, adancime + 1, adancime_maxima)}

def prezice_arbore(arbore, x):
    if arbore["tip"] == "frunza":
        return arbore["clasa"]
    if x[arbore["coloana"]] <= arbore["prag"]:
        return prezice_arbore(arbore["stanga"], x)
    return prezice_arbore(arbore["dreapta"], x)

X = [[1, 1], [1, 0], [0, 1], [0, 0]]
y = [1, 1, 0, 0]
arbore = construieste_arbore(X, y, adancime_maxima=2)
print(arbore)
print(prezice_arbore(arbore, [1, 0]))
````

# 8. K-Means

K-Means grupeaza observatiile in `k` clustere.

## Pseudocod

```text
Algoritm KMeans
Intrare: date X, numar clustere k, iteratii
Iesire: centroizi, etichete

Alege primii k centroizi
Pentru fiecare iteratie:
    Pentru fiecare punct:
        calculeaza distanta fata de fiecare centroid
        atribuie punctul celui mai apropiat centroid
    Pentru fiecare cluster:
        recalculeaza centroidul ca medie a punctelor atribuite
Returneaza centroizii si etichetele
```

## Implementare Python

````python
def medie_puncte(puncte):
    dim = len(puncte[0])
    centroid = [0] * dim
    for punct in puncte:
        for j in range(dim):
            centroid[j] += punct[j]
    for j in range(dim):
        centroid[j] /= len(puncte)
    return centroid

def kmeans(X, k, iteratii=10):
    centroizi = [X[i][:] for i in range(k)]
    etichete = [0] * len(X)
    for _ in range(iteratii):
        for i in range(len(X)):
            distante = []
            for c in centroizi:
                distante.append(distanta_euclidiana(X[i], c))
            etichete[i] = distante.index(min(distante))
        for cluster in range(k):
            puncte_cluster = []
            for i in range(len(X)):
                if etichete[i] == cluster:
                    puncte_cluster.append(X[i])
            if len(puncte_cluster) > 0:
                centroizi[cluster] = medie_puncte(puncte_cluster)
    return centroizi, etichete

clienti = [[25, 2000], [30, 2500], [28, 2300], [50, 8000], [55, 8500], [52, 8200]]
centroizi, etichete = kmeans(clienti, 2, iteratii=20)
print("Centroizi:", centroizi)
print("Etichete:", etichete)
````

# 9. Gradient Descent

Gradient Descent minimizeaza o functie prin pasi succesivi in directia opusa gradientului.

Exemplu: `f(x) = x^2`, iar `f'(x) = 2x`.

## Pseudocod

```text
Algoritm GradientDescent
Intrare: valoare initiala x, rata alpha, iteratii
Iesire: x optimizat

Pentru fiecare iteratie:
    gradient <- 2*x
    x <- x - alpha * gradient
Returneaza x
```

## Implementare Python

````python
def gradient_descent_x2(x_initial, alpha=0.1, iteratii=50):
    x = x_initial
    for _ in range(iteratii):
        gradient = 2 * x
        x = x - alpha * gradient
    return x

def cost(x):
    return (x - 5) ** 2

def gradient_cost(x):
    return 2 * (x - 5)

def optimizeaza_cost(x_initial, alpha=0.1, iteratii=50):
    x = x_initial
    for _ in range(iteratii):
        x -= alpha * gradient_cost(x)
    return x

print(gradient_descent_x2(10))
x_opt = optimizeaza_cost(0)
print("x optim:", x_opt)
print("cost:", cost(x_opt))
````

# 10. Programare liniara prin cautare pe grila

Model:

```text
max z = c1*x1 + c2*x2
```

## Pseudocod

```text
Algoritm CautareGrilaLP
Intrare: coeficienti obiectiv, constrangeri, limita maxima
Iesire: cea mai buna solutie gasita

best_val <- -infinit
best_x <- None
Pentru x1 de la 0 la limita:
    Pentru x2 de la 0 la limita:
        Daca toate constrangerile sunt respectate:
            val <- c1*x1 + c2*x2
            Daca val > best_val:
                best_val <- val
                best_x <- (x1, x2)
Returneaza best_x, best_val
```

## Implementare Python

````python
def respecta_constrangeri(x1, x2, A, b):
    for i in range(len(A)):
        if A[i][0] * x1 + A[i][1] * x2 > b[i]:
            return False
    return True

def programare_liniara_grila(c, A, b, limita=20, pas=1):
    best_val = -10**18
    best_x = None
    x1 = 0
    while x1 <= limita:
        x2 = 0
        while x2 <= limita:
            if respecta_constrangeri(x1, x2, A, b):
                val = c[0] * x1 + c[1] * x2
                if val > best_val:
                    best_val = val
                    best_x = (x1, x2)
            x2 += pas
        x1 += pas
    return best_x, best_val

c = [5, 4]
A = [[6, 4], [1, 2], [-1, 1]]
b = [24, 6, 1]
print(programare_liniara_grila(c, A, b, limita=10, pas=0.1))
````

# 11. Metoda Simplex simplificata

Forma tratata:

```text
max c*x
A*x <= b
x >= 0
```

## Pseudocod

```text
Algoritm SimplexSimplificat
Intrare: A, b, c
Iesire: solutie optima

Construieste tabelul simplex cu variabile slack
Cat timp exista coeficient pozitiv in linia obiectivului:
    alege coloana pivot cu cel mai mare coeficient pozitiv
    calculeaza rapoartele b_i / a_i_pivot
    alege linia pivot cu raport minim pozitiv
    normalizeaza linia pivot
    elimina coloana pivot din celelalte linii
Returneaza valorile variabilelor si optimul
```

## Implementare Python

````python
def simplex(A, b, c):
    m = len(A)
    n = len(c)
    tabel = []
    for i in range(m):
        rand = A[i][:] + [0] * m + [b[i]]
        rand[n + i] = 1
        tabel.append(rand)
    obiectiv = c[:] + [0] * m + [0]
    tabel.append(obiectiv)
    while True:
        ultima = tabel[-1]
        coloana_pivot = None
        maxim = 0
        for j in range(len(ultima) - 1):
            if ultima[j] > maxim:
                maxim = ultima[j]
                coloana_pivot = j
        if coloana_pivot is None:
            break
        rapoarte = []
        for i in range(m):
            if tabel[i][coloana_pivot] > 0:
                rapoarte.append((tabel[i][-1] / tabel[i][coloana_pivot], i))
        if not rapoarte:
            raise Exception("Problema nelimitata")
        linie_pivot = min(rapoarte)[1]
        pivot = tabel[linie_pivot][coloana_pivot]
        for j in range(len(tabel[linie_pivot])):
            tabel[linie_pivot][j] /= pivot
        for i in range(len(tabel)):
            if i != linie_pivot:
                factor = tabel[i][coloana_pivot]
                for j in range(len(tabel[i])):
                    tabel[i][j] -= factor * tabel[linie_pivot][j]
    solutie = [0] * n
    for j in range(n):
        coloana = [tabel[i][j] for i in range(m)]
        if coloana.count(1) == 1 and coloana.count(0) == m - 1:
            linie = coloana.index(1)
            solutie[j] = tabel[linie][-1]
    optim = -tabel[-1][-1]
    return solutie, optim

A = [[6, 4], [1, 2]]
b = [24, 6]
c = [5, 4]
print(simplex(A, b, c))
````

# 12. Problema transportului

## 12.1 Metoda coltului nord-vestic

## Pseudocod

```text
Algoritm ColtNordVestic
Intrare: oferta, cerere
Iesire: plan de transport

i <- 0; j <- 0
Cat timp i < numar_surse si j < numar_destinatii:
    cantitate <- min(oferta[i], cerere[j])
    x[i][j] <- cantitate
    oferta[i] <- oferta[i] - cantitate
    cerere[j] <- cerere[j] - cantitate
    Daca oferta[i] = 0: i <- i + 1
    Daca cerere[j] = 0: j <- j + 1
Returneaza x
```

## Implementare Python

````python
def colt_nord_vestic(oferta, cerere):
    oferta = oferta[:]
    cerere = cerere[:]
    m = len(oferta)
    n = len(cerere)
    plan = [[0 for _ in range(n)] for _ in range(m)]
    i = 0
    j = 0
    while i < m and j < n:
        cantitate = min(oferta[i], cerere[j])
        plan[i][j] = cantitate
        oferta[i] -= cantitate
        cerere[j] -= cantitate
        if oferta[i] == 0:
            i += 1
        if j < n and cerere[j] == 0:
            j += 1
    return plan

def cost_transport(plan, costuri):
    total = 0
    for i in range(len(plan)):
        for j in range(len(plan[0])):
            total += plan[i][j] * costuri[i][j]
    return total

oferta = [20, 30]
cerere = [10, 25, 15]
costuri = [[2, 3, 1], [5, 4, 8]]
plan = colt_nord_vestic(oferta, cerere)
print(plan)
print("Cost:", cost_transport(plan, costuri))
````

## 12.2 Metoda costului minim

```text
Algoritm CostMinim
Intrare: oferta, cerere, costuri

Cat timp exista oferta si cerere:
    gaseste celula disponibila cu cost minim

```

## Implementare Python

````python
def metoda_costului_minim(oferta, cerere, costuri):
    oferta = oferta[:]
    cerere = cerere[:]
    m = len(oferta)
    n = len(cerere)
    plan = [[0 for _ in range(n)] for _ in range(m)]
    while sum(oferta) > 0 and sum(cerere) > 0:
        best = None
        for i in range(m):
            for j in range(n):
                if oferta[i] > 0 and cerere[j] > 0:
                    if best is None or costuri[i][j] < costuri[best[0]][best[1]]:
                        best = (i, j)
        i, j = best
        cantitate = min(oferta[i], cerere[j])
        plan[i][j] = cantitate
        oferta[i] -= cantitate
        cerere[j] -= cantitate
    return plan

plan2 = metoda_costului_minim(oferta, cerere, costuri)
print(plan2)
print("Cost:", cost_transport(plan2, costuri))
````

# 13. Algoritmi Monte Carlo

Monte Carlo foloseste esantionare aleatoare pentru estimarea unor marimi.

## Pseudocod pentru estimarea lui pi

```text
Algoritm EstimarePiMonteCarlo
Intrare: numar simulari n
Iesire: estimare pi

inside <- 0
Pentru iteratia 1..n:
    genereaza x aleator in [0,1]
    genereaza y aleator in [0,1]
    daca x^2 + y^2 <= 1:
        inside <- inside + 1
pi <- 4 * inside / n
Returneaza pi
```

## Implementare Python

````python
import random

def estimeaza_pi(n=10000):
    inside = 0
    for _ in range(n):
        x = random.random()
        y = random.random()
        if x*x + y*y <= 1:
            inside += 1
    return 4 * inside / n

print(estimeaza_pi(10000))

def simulare_portofoliu(capital_initial, randament_mediu, volatilitate, ani, simulari):
    rezultate = []
    for _ in range(simulari):
        capital = capital_initial
        for _ in range(ani):
            randament = random.gauss(randament_mediu, volatilitate)
            capital *= (1 + randament)
        rezultate.append(capital)
    return rezultate

rezultate = simulare_portofoliu(10000, 0.07, 0.15, 10, 1000)
print("Valoare medie finala:", media(rezultate))
print("Valoare minima:", min(rezultate))
````

# 14. Algoritm genetic

Elemente: populatie, fitness, selectie, crossover si mutatie.

## Pseudocod

```text
Algoritm Genetic
Intrare: dimensiune populatie, numar generatii
Iesire: cel mai bun individ

Genereaza populatia initiala
Pentru fiecare generatie:
    calculeaza fitness pentru fiecare individ
    selecteaza parinti
    aplica crossover
    aplica mutatie
    formeaza noua populatie
Returneaza individul cu fitness maxim
```

## Implementare Python

````python
def fitness(x):
    return x * x

def selectie(populatie):
    a = random.choice(populatie)
    b = random.choice(populatie)
    return a if fitness(a) > fitness(b) else b

def crossover(a, b):
    masca = random.randint(1, 30)
    copil = (a & masca) | (b & (~masca & 31))
    return copil

def mutatie(x, rata=0.1):
    if random.random() < rata:
        bit = random.randint(0, 4)
        x = x ^ (1 << bit)
    return x

def algoritm_genetic(dim_pop=10, generatii=30):
    populatie = [random.randint(0, 31) for _ in range(dim_pop)]
    for _ in range(generatii):
        noua_populatie = []
        for _ in range(dim_pop):
            p1 = selectie(populatie)
            p2 = selectie(populatie)
            copil = crossover(p1, p2)
            copil = mutatie(copil)
            noua_populatie.append(copil)
        populatie = noua_populatie
    best = max(populatie, key=fitness)
    return best, fitness(best)

print(algoritm_genetic())
````

# 15. Programare dinamica: problema rucsacului 0/1

Fiecare obiect poate fi selectat sau neselectat.

## Pseudocod

```text
Algoritm KnapsackDP
Intrare: greutati, valori, capacitate
Iesire: valoare maxima

Creeaza tabel dp cu dimensiunea (n+1) x (capacitate+1)
Pentru i de la 1 la n:
    Pentru w de la 0 la capacitate:
        dp[i][w] <- dp[i-1][w]
        Daca greutate[i-1] <= w:
            dp[i][w] <- max(dp[i][w], dp[i-1][w-greutate[i-1]] + valoare[i-1])
Returneaza dp[n][capacitate]
```

## Implementare Python

````python
def knapsack_dp(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            dp[i][w] = dp[i - 1][w]
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
    return dp[n][capacity]

costuri = [2, 3, 4, 5]
profituri = [30, 40, 50, 60]
buget = 5
print(knapsack_dp(costuri, profituri, buget))
````

# 16. Simulare si teoria cozilor

Simularea cozilor este utila pentru banci, supermarketuri, spitale si centre de suport.

## Pseudocod

```text
Algoritm SimulareCoada
Intrare: numar clienti, rata sosire, rata servire
Iesire: timpi de asteptare

moment_curent_sosire <- 0
moment_liber_server <- 0
Pentru fiecare client:
    genereaza timpul pana la urmatoarea sosire
    actualizeaza momentul sosirii
    genereaza timpul de servire
    inceput_servire <- max(moment_sosire, moment_liber_server)
    asteptare <- inceput_servire - moment_sosire
    moment_liber_server <- inceput_servire + timp_servire
Returneaza timpii de asteptare
```

## Implementare Python

````python
def simulare_coada(numar_clienti, rata_sosire, rata_servire):
    moment_sosire = 0
    moment_liber_server = 0
    asteptari = []
    for _ in range(numar_clienti):
        timp_pana_la_sosire = random.expovariate(rata_sosire)
        timp_servire = random.expovariate(rata_servire)
        moment_sosire += timp_pana_la_sosire
        inceput_servire = max(moment_sosire, moment_liber_server)
        asteptare = inceput_servire - moment_sosire
        moment_liber_server = inceput_servire + timp_servire
        asteptari.append(asteptare)
    return asteptari

asteptari = simulare_coada(100, rata_sosire=0.2, rata_servire=0.25)
print("Asteptare medie:", media(asteptari))
print("Asteptare maxima:", max(asteptari))
````

# 17. Serii temporale

Aplicatii: inflatie, PIB, vanzari, curs valutar si cerere de energie.

## 17.1 Media mobila

## Pseudocod

```text
Algoritm MediaMobila
Intrare: serie x, fereastra k
Iesire: serie netezita

Pentru fiecare pozitie i:
    Daca i < k-1:
        adauga None
    Altfel:
        calculeaza media ultimelor k valori
Returneaza seria netezita
```

## Implementare Python

````python
def media_mobila(x, k):
    rezultat = []
    for i in range(len(x)):
        if i < k - 1:
            rezultat.append(None)
        else:
            fereastra = x[i-k+1:i+1]
            rezultat.append(media(fereastra))
    return rezultat

def netezire_exponentiala(x, alpha):
    rezultat = [x[0]]
    for t in range(1, len(x)):
        s = alpha * x[t] + (1 - alpha) * rezultat[-1]
        rezultat.append(s)
    return rezultat

vanzari_lunare = [100, 120, 130, 125, 140, 150, 160]
print("Media mobila:", media_mobila(vanzari_lunare, 3))
print("Netezire exponentiala:", netezire_exponentiala(vanzari_lunare, 0.4))
````
