# Partea I – Arbori binari


## 1. Ideea de baza

Un **arbore binar** este o structura de date ierarhica formata din noduri.  
Fiecare nod poate avea:
- cel mult un copil stang;
- cel mult un copil drept.

Aceasta este diferenta principala fata de structurile liniare, cum sunt vectorii sau listele.  
Intr-un arbore, informatia nu este asezata una dupa alta, ci se ramifica.

Arborii binari sunt importanti deoarece apar in:
- organizarea ierarhica a datelor;
- cautare;
- parsarea expresiilor;
- multe algoritmi recursivi;
- structuri mai avansate, cum sunt arborii binari de cautare.

### Termeni importanti
- **radacina** – primul nod al arborelui;
- **copil** – nod aflat direct sub alt nod;
- **parinte** – nod care are unul sau mai multi copii;
- **frunza** – nod fara copii;
- **subarbore** – o parte a arborelui care poate fi privita ca arbore separat;
- **nivel** – distanta fata de radacina;
- **inaltime** – lungimea maxima a unui drum de la radacina la o frunza.


## 2. Reprezentare intuitiva

Vom lucra cu urmatorul arbore:

```text
        10
       /  \
      5    15
     / \   / \
    2   7 12  20
```

Observatii:
- `10` este radacina;
- `5` si `15` sunt copiii radacinii;
- `2`, `7`, `12`, `20` sunt frunze;
- subarborele stang al radacinii este arborele care incepe cu `5`;
- subarborele drept al radacinii este arborele care incepe cu `15`.

Este util sa privesti arborele ca pe o structura care separa informatia in zone:
- ce este in stanga;
- ce este in dreapta;
- cum se combina aceste parti.


## 3. Reprezentarea unui nod

Fiecare nod trebuie sa poata retine:
- o valoare;
- o legatura catre copilul stang;
- o legatura catre copilul drept.

Aceasta este forma minima necesara pentru a construi un arbore binar.


````python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
````

## 4. Construirea unui arbore

Mai jos construim manual arborele din exemplul anterior.


````python
root = Node(10)

root.left = Node(5)
root.right = Node(15)

root.left.left = Node(2)
root.left.right = Node(7)

root.right.left = Node(12)
root.right.right = Node(20)

print("Radacina:", root.val)
print("Copil stang:", root.left.val)
print("Copil drept:", root.right.val)
````

## 5. Ce inseamna parcurgerea unui arbore?

A parcurge un arbore inseamna a vizita nodurile intr-o anumita ordine.

Cele mai importante trei parcurgeri sunt:
- **inorder**: stanga → radacina → dreapta
- **preorder**: radacina → stanga → dreapta
- **postorder**: stanga → dreapta → radacina

Aceste ordini sunt esentiale.  
Acelasi arbore poate produce secvente diferite de valori in functie de regula de parcurgere.


## 6. Parcurgerea inorder

La inorder:
1. mergem in subarborele stang;
2. procesam nodul curent;
3. mergem in subarborele drept.

Aceasta parcurgere este foarte importanta deoarece, in anumite tipuri de arbori, produce valorile in ordine crescatoare.


````python
def inorder(node):
    if node is not None:
        inorder(node.left)
        print(node.val, end=" ")
        inorder(node.right)

inorder(root)
````

### Explicatie pas cu pas pentru inorder

Pornim din `10`.
- mai intai mergem in stanga, la `5`;
- din `5` mergem in stanga, la `2`;
- afisam `2`;
- revenim la `5` si il afisam;
- mergem la `7` si il afisam;
- revenim la `10` si il afisam;
- apoi mergem in dreapta, la `15`;
- afisam `12`, apoi `15`, apoi `20`.

Ordinea finala este:
```text
2 5 7 10 12 15 20
```

Aceasta explicatie trebuie urmarita pe desen. La inceput, desenul este esential.


## 7. Parcurgerea preorder

La preorder:
1. procesam nodul curent;
2. mergem in stanga;
3. mergem in dreapta.

Aceasta parcurgere este utila in probleme in care vrem sa tratam mai intai radacina si apoi ramurile ei.


````python
def preorder(node):
    if node is not None:
        print(node.val, end=" ")
        preorder(node.left)
        preorder(node.right)

preorder(root)
````

## 8. Parcurgerea postorder

La postorder:
1. mergem in stanga;
2. mergem in dreapta;
3. procesam nodul curent.

Aceasta ordine este utila in situatii in care copiii trebuie tratati inaintea parintelui.


````python
def postorder(node):
    if node is not None:
        postorder(node.left)
        postorder(node.right)
        print(node.val, end=" ")

postorder(root)
````

## 9. Comparatie intre cele trei parcurgeri

Pentru acelasi arbore, cele trei parcurgeri produc ordine diferite:

- **inorder**: stanga, nod, dreapta
- **preorder**: nod, stanga, dreapta
- **postorder**: stanga, dreapta, nod

Un exercitiu foarte util este sa desenezi arborele si sa simulezi manual fiecare parcurgere.  
Aceasta activitate formeaza intuitia necesara pentru probleme mai dificile.


## 10. Exercitiu rezolvat: numarul de noduri

### Enunt
Sa se determine cate noduri exista in arbore.

### Idee
Pentru fiecare nod:
- il numaram pe el;
- numaram nodurile din stanga;
- numaram nodurile din dreapta.

Formula este:
```text
1 + numar_stanga + numar_dreapta
```


````python
def count_nodes(node):
    if node is None:
        return 0
    return 1 + count_nodes(node.left) + count_nodes(node.right)

print(count_nodes(root))
````

### Explicatie
Daca nodul nu exista, nu avem nimic de numarat, deci intoarcem `0`.

Daca nodul exista, intoarcem:
- `1` pentru nodul curent;
- plus rezultatul obtinut din stanga;
- plus rezultatul obtinut din dreapta.

Acesta este unul dintre cele mai importante modele de gandire pe arbori:
**problema de la un nod se descompune in aceeasi problema pe subarborele stang si pe subarborele drept**.


## 11. Exercitiu rezolvat: inaltimea arborelui

### Enunt
Sa se determine inaltimea arborelui.

### Idee
Inaltimea urmareste drumul cel mai lung pana la o frunza.

Prin urmare:
- daca nodul nu exista, intoarcem `0`;
- altfel, intoarcem `1 + max(inaltime_stanga, inaltime_dreapta)`.


````python
def height(node):
    if node is None:
        return 0
    return 1 + max(height(node.left), height(node.right))

print(height(root))
````

### Explicatie
Folosim `max` deoarece nu ne intereseaza orice drum, ci drumul cel mai lung.

Daca un subarbore are inaltime 2 si celalalt are inaltime 3, atunci nodul curent va avea inaltime `1 + 3 = 4`.


## 12. Exercitiu rezolvat: numarul de frunze

### Enunt
Sa se determine cate frunze are arborele.

### Idee
Un nod este frunza daca:
- exista;
- nu are copil stang;
- nu are copil drept.


````python
def count_leaves(node):
    if node is None:
        return 0

    if node.left is None and node.right is None:
        return 1

    return count_leaves(node.left) + count_leaves(node.right)

print(count_leaves(root))
````

## 13. Exercitiu rezolvat: suma valorilor din arbore

### Enunt
Sa se calculeze suma tuturor valorilor.

### Idee
Pentru fiecare nod:
- luam valoarea lui;
- adaugam suma din stanga;
- adaugam suma din dreapta.


````python
def sum_nodes(node):
    if node is None:
        return 0
    return node.val + sum_nodes(node.left) + sum_nodes(node.right)

print(sum_nodes(root))
````

## 14. Exercitiu rezolvat: cautarea unei valori

### Enunt
Sa se determine daca o valoare exista in arbore.

### Idee
Comparam valoarea cautata cu nodul curent.
Daca nu este aici, cautam in stanga si in dreapta.


````python
def contains(node, x):
    if node is None:
        return False

    if node.val == x:
        return True

    return contains(node.left, x) or contains(node.right, x)

print(contains(root, 7))
print(contains(root, 100))
````

## 15. Exercitiu rezolvat: afisarea frunzelor

### Enunt
Sa se afiseze toate frunzele arborelui.

### Idee
Parcurgem arborele si afisam doar nodurile care nu au copii.


````python
def print_leaves(node):
    if node is None:
        return

    if node.left is None and node.right is None:
        print(node.val, end=" ")
        return

    print_leaves(node.left)
    print_leaves(node.right)

print_leaves(root)
````

# Partea a II-a – Grafuri

## 20. Ideea de baza

Un **graf** este o structura formata din:
- noduri;
- muchii care leaga nodurile intre ele.

Grafurile sunt folosite pentru a descrie relatii.

Exemple:
- orase si drumuri;
- persoane si relatii de prietenie;
- calculatoare si conexiuni;
- statii si trasee;
- pagini web si linkuri.


## 21. Reprezentare intuitiva

Un graf mic poate fi vazut astfel:

```text
A -> B, C
B -> D, E
C -> F
D -> nimic
E -> F
F -> nimic
```

Aici:
- din `A` putem merge la `B` si `C`;
- din `B` putem merge la `D` si `E`;
- din `C` putem merge la `F`.

Aceasta structura este suficienta pentru a intelege primele parcurgeri.


## 22. Reprezentarea prin lista de adiacenta

O lista de adiacenta spune, pentru fiecare nod, care sunt vecinii sai.

Este una dintre cele mai simple si mai utile reprezentari pentru grafuri.


````python
graf = {
    "A": ["B", "C"],
    "B": ["D", "E"],
    "C": ["F"],
    "D": [],
    "E": ["F"],
    "F": []
}

print(graf)
````

## 23. Ce inseamna parcurgerea unui graf?

A parcurge un graf inseamna a porni dintr-un nod si a vizita nodurile accesibile din el.

Cele mai importante doua strategii sunt:
- **BFS** – parcurgere pe niveluri;
- **DFS** – parcurgere in adancime.

Aceste doua metode apar foarte frecvent in probleme de concurs, examene si programare practica.


## 24. BFS – Breadth-First Search

BFS viziteaza nodurile pe niveluri:
- mai intai nodul de start;
- apoi vecinii lui;
- apoi vecinii vecinilor.

BFS foloseste o **coada**.

De ce?
Pentru ca vrem sa procesam nodurile in ordinea in care au fost descoperite.


````python
from collections import deque

def bfs(g, start):
    q = deque([start])
    viz = set([start])

    while q:
        node = q.popleft()
        print(node, end=" ")

        for vecin in g[node]:
            if vecin not in viz:
                viz.add(vecin)
                q.append(vecin)

bfs(graf, "A")
````

### Explicatie pas cu pas pentru BFS

Pornim din `A`.
- initial, coada contine `A`;
- scoatem `A`, apoi adaugam `B` si `C`;
- scoatem `B`, apoi adaugam `D` si `E`;
- scoatem `C`, apoi adaugam `F`;
- continuam pana se goleste coada.

Ordinea este:
```text
A B C D E F
```

BFS este foarte important atunci cand vrem sa gasim drumul minim in numar de pasi intr-un graf neponderat.


## 25. DFS – Depth-First Search

DFS merge in adancime:
- pornim dintr-un nod;
- alegem un vecin;
- mergem cat mai departe;
- apoi revenim.

DFS poate fi scris:
- recursiv;
- iterativ cu o stiva.

Pentru inceput, varianta recursiva este mai usor de inteles.


````python
def dfs(g, node, viz):
    if node in viz:
        return

    viz.add(node)
    print(node, end=" ")

    for vecin in g[node]:
        dfs(g, vecin, viz)

dfs(graf, "A", set())
````

## 26. Diferenta dintre BFS si DFS

### BFS
- parcurgere pe niveluri;
- foloseste coada;
- este foarte bun pentru distante minime.

### DFS
- parcurgere in adancime;
- foloseste recursivitate sau stiva;
- este foarte bun pentru explorare completa.

Este important sa poti recunoaste ce tip de parcurgere cere problema.


## 27. Exercitiu rezolvat: ordinea BFS intr-o lista

### Enunt
Sa se modifice BFS astfel incat sa intoarca ordinea de parcurgere intr-o lista.


````python
def bfs_list(g, start):
    q = deque([start])
    viz = set([start])
    ordine = []

    while q:
        node = q.popleft()
        ordine.append(node)

        for vecin in g[node]:
            if vecin not in viz:
                viz.add(vecin)
                q.append(vecin)

    return ordine

print(bfs_list(graf, "A"))
````

## 28. Exercitiu rezolvat: ordinea DFS intr-o lista

### Enunt
Sa se modifice DFS astfel incat sa intoarca lista nodurilor vizitate.


````python
def dfs_list(g, start):
    ordine = []
    viz = set()

    def aux(node):
        if node in viz:
            return
        viz.add(node)
        ordine.append(node)

        for vecin in g[node]:
            aux(vecin)

    aux(start)
    return ordine

print(dfs_list(graf, "A"))
````

## 29. Exercitiu rezolvat: exista drum intre doua noduri?

### Enunt
Sa se determine daca exista drum intre `start` si `end`.


````python
def exista_drum(g, start, end):
    stack = [start]
    viz = set()

    while stack:
        node = stack.pop()

        if node == end:
            return True

        if node not in viz:
            viz.add(node)
            for vecin in g[node]:
                stack.append(vecin)

    return False

print(exista_drum(graf, "A", "F"))
print(exista_drum(graf, "C", "D"))
````

### Explicatie
Pornim din `start` si exploram tot ce putem atinge.

Daca, la un moment dat, ajungem la `end`, inseamna ca exista drum.  
Daca explorarea se termina fara sa gasim nodul, atunci nu exista drum.


## 30. Exercitiu rezolvat: distante minime cu BFS

### Enunt
Sa se calculeze distanta minima de la nodul de start la toate celelalte noduri.

### Idee
In BFS, prima data cand descoperim un nod, stim deja distanta minima pana la el.


````python
def distante_bfs(g, start):
    q = deque([start])
    dist = {start: 0}

    while q:
        node = q.popleft()

        for vecin in g[node]:
            if vecin not in dist:
                dist[vecin] = dist[node] + 1
                q.append(vecin)

    return dist

print(distante_bfs(graf, "A"))
````

## 31. Exercitiu rezolvat: toate nodurile accesibile

### Enunt
Sa se determine toate nodurile accesibile dintr-un nod dat.


````python
def accesibile(g, start):
    return bfs_list(g, start)

print(accesibile(graf, "B"))
print(accesibile(graf, "C"))
````

## 32. Exercitiu rezolvat: componente conexe

Vom folosi un graf neorientat simplu.


````python
graf2 = {
    "A": ["B"],
    "B": ["A"],
    "C": ["D"],
    "D": ["C"],
    "E": []
}

def dfs_comp(g, node, viz):
    viz.add(node)
    for vecin in g[node]:
        if vecin not in viz:
            dfs_comp(g, vecin, viz)

def componente_conexe(g):
    viz = set()
    count = 0

    for node in g:
        if node not in viz:
            dfs_comp(g, node, viz)
            count += 1

    return count

print(componente_conexe(graf2))
````

## 33. Explicatie pentru componente conexe

De fiecare data cand gasim un nod nevizitat:
- pornim o parcurgere din el;
- marcam toate nodurile din acea zona;
- crestem contorul de componente.

La final, numarul de porniri distincte este numarul componentelor conexe.
